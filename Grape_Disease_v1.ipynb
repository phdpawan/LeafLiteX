### Grape Disease Detection - Final Code Run

#### MobileNetV3Large

```python
import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.metrics import classification_report, confusion_matrix
import os
import shutil
from collections import defaultdict
import math
from collections import Counter
import pandas as pd

# Random seed setup
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

# Dataset path local/colab
# Colab
DATASET_PATH = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated")
# Local
# DATASET_PATH = Path("../Datasets/plant-village-dataset-updated-aggregated")

# Change crop name here
CROP_NAME = "Grape"
CROP_PATH = DATASET_PATH / CROP_NAME
IMG_SIZE = (224, 224)
# CPU
# BATCH_SIZE = 32
# GPU
BATCH_SIZE = 64
EPOCHS = 30
PRE_TRAINED_MODEL = "MobileNetV3Large_TrainTestVal"
DROPOUT = 0.3
LR_RATE = 1e-4

OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")
OUT_DIR.mkdir(parents=True, exist_ok=True)


# Data Loader (70% / 15% / 15%)
# Count per class in each split
def count_per_class(file_paths, class_names):
    labels = [Path(f).parent.name for f in file_paths]
    counts = Counter(labels)
    return {cls: counts.get(cls, 0) for cls in class_names}

# Gather file list to compute exact counts
all_files = []
for root, _, files in os.walk(CROP_PATH):
    for f in files:
        if f.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp")):
            all_files.append(Path(root) / f)
N = len(all_files)
if N == 0:
    raise RuntimeError(f"No images found under {CROP_PATH}")

train_count = int(N * 0.70)
val_count = int(N * 0.15)
# Validate sum
test_count = N - train_count - val_count

print(f"Total images: {N} -> train: {train_count}, val: {val_count}, test: {test_count}")

# Use image_dataset_from_directory with validation_split=0.3:
rest_split = 0.3

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    CROP_PATH,
    validation_split=rest_split,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True
)

rest_ds = tf.keras.preprocessing.image_dataset_from_directory(
    CROP_PATH,
    validation_split=rest_split,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,   # batched
    shuffle=True
)

class_names = train_ds.class_names
num_classes = len(class_names)

# Print the count of images per disease for Train, Test, and Val
train_files = train_ds.file_paths
rest_files = [p for p in rest_ds.file_paths]
val_files = rest_files[:val_count]
test_files = rest_files[val_count:val_count+test_count]

train_counts = count_per_class(train_files, class_names)
val_counts = count_per_class(val_files, class_names)
test_counts = count_per_class(test_files, class_names)

# Create DataFrame for thesis table
df = pd.DataFrame({
    "Train": train_counts,
    "Val": val_counts,
    "Test": test_counts
})
print("\n Per-class image distribution:\n")
print(df)
column_sums = df.sum()

print("\n--------------------")
print("Sum of each column:")
print(column_sums)

# Flatten rest_ds and split deterministically by counts
rest_ds = rest_ds.unbatch()
val_ds = rest_ds.take(val_count)
test_ds = rest_ds.skip(val_count).take(test_count)

# Rebatch val & test for training pipeline; we'll also create finite eval datasets below
val_ds = val_ds.batch(BATCH_SIZE)
test_ds = test_ds.batch(BATCH_SIZE)

# Derive class_names from train_ds
print("\n--------------------")
print(f"✅ Training {CROP_NAME} model with {num_classes} classes")
print("Classes:", class_names)
train_ds_len = tf.data.experimental.cardinality(train_ds)
print(f"Total images: {N} -> train data per batch: {train_ds_len.numpy()}")

# Save class labels
with open(f"{CROP_NAME.lower()}_disease_labels.txt", "w") as f:
    for name in class_names:
        f.write(name + "\n")

# Data augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
    tf.keras.layers.RandomContrast(0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
    tf.keras.layers.GaussianNoise(0.05),
    tf.keras.layers.RandomBrightness(factor=0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
])

AUTOTUNE = tf.data.AUTOTUNE


# Pipelines

# TRAINING pipeline: cache -> shuffle -> augment -> repeat -> prefetch
train_ds = (
    train_ds
    .cache()
    .shuffle(1000)
    .map(lambda x, y: (data_augmentation(x, training=True), y),
         num_parallel_calls=AUTOTUNE)
    .repeat()   # <-- ADD repeat here so fit() can run many epochs
    .prefetch(buffer_size=AUTOTUNE)
)

# VALIDATION for training (use a repeated version for validation during fit)
val_ds_for_fit = (
    val_ds
    .cache()
    .repeat()
    .prefetch(buffer_size=AUTOTUNE)
)

# VAL pipeline used for final evaluation remains finite (val_ds)
# TEST pipeline remains finite
test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE)

# compute steps for training loops
steps_per_epoch = max(1, train_count // BATCH_SIZE)
validation_steps = max(1, val_count // BATCH_SIZE)
print(f"steps_per_epoch={steps_per_epoch}, validation_steps={validation_steps}")

# Model Building
base_model = tf.keras.applications.MobileNetV3Large(
    input_shape=IMG_SIZE + (3,),
    include_top=False,
    weights="imagenet",
    include_preprocessing=True
)
base_model.trainable = False  # freeze backbone initially

# Input -> backbone -> head (we reuse base_model.input/output)
# x = base_model.output
# x = tf.keras.layers.GlobalAveragePooling2D()(x)
# x = tf.keras.layers.Dropout(0.3)(x)
# outputs = tf.keras.layers.Dense(num_classes, activation="softmax", name="predictions")(x)

x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(256, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(LR_RATE))(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dropout(DROPOUT)(x)
outputs = tf.keras.layers.Dense(num_classes, activation="softmax", name="predictions")(x)

model = tf.keras.Model(base_model.input, outputs, name=f"{PRE_TRAINED_MODEL}_{CROP_NAME}")

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# Training
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss", patience=10, restore_best_weights=True
)

checkpoint = tf.keras.callbacks.ModelCheckpoint(
    f"best_{CROP_NAME.lower()}_disease_model_{PRE_TRAINED_MODEL}.keras",
    save_best_only=True,
    monitor="val_accuracy",
    mode="max"
)

history = model.fit(
    train_ds,
    validation_data=val_ds_for_fit,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# Fine-tuning: unfreeze some top layers of base_model
base_model.trainable = True
fine_tune_at = int(len(base_model.layers) * 0.75)
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_fine = model.fit(
    train_ds,
    validation_data=val_ds_for_fit,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# Evaluate on validation
results = model.evaluate(val_ds)
print("Metrics - ")
for name, value in zip(model.metrics_names, results):
    print(f"{name}: {value:.4f}")

# Plots: Accuracy & Loss
plt.figure(figsize=(10, 6))
plt.plot(history.history.get('accuracy', []) + history_fine.history.get('accuracy', []), label="Training Accuracy")
plt.plot(history.history.get('val_accuracy', []) + history_fine.history.get('val_accuracy', []), label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title(f"{CROP_NAME} - Training & Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(history.history.get('loss', []) + history_fine.history.get('loss', []), label="Training Loss")
plt.plot(history.history.get('val_loss', []) + history_fine.history.get('val_loss', []), label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title(f"{CROP_NAME} - Training & Validation Loss")
plt.legend()
plt.grid(True)
plt.show()

# Validation Evaluation: Classification Report & Confusion Matrix on VAL set
# ==============================
y_true = []
y_pred = []

for batch_images, batch_labels in val_ds:
    preds = model.predict(batch_images, verbose=0)
    y_true.extend(batch_labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

# Safety: ensure lengths align
if len(class_names) != (int(max(max(y_true, default=0), max(y_pred, default=0))) + 1):
    print("Warning: class_names length does not match labels. Recomputing class_names from folders.")
    class_names = sorted([d.name for d in CROP_PATH.iterdir() if d.is_dir()])
    num_classes = len(class_names)
    print("New class_names:", class_names)

y_true = np.array(y_true, dtype=np.int64)
y_pred = np.array(y_pred, dtype=np.int64)

acc = float(np.mean(y_true == y_pred))
print(f"\n Overall accuracy: {acc*100:.2f}%")

print("Validation Classification Report:")
print(classification_report(y_true, y_pred,labels=list(range(num_classes)), target_names=class_names, zero_division=0))

cm = confusion_matrix(y_true, y_pred, labels=range(len(class_names)))
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"{CROP_NAME} - Validation Confusion Matrix")
plt.show()

# Testing Evaluation: classification report & confusion matrix on TEST set
y_true_test = []
y_pred_test = []

for batch_images, batch_labels in test_ds:
    preds = model.predict(batch_images, verbose=0)
    y_true_test.extend(batch_labels.numpy())
    y_pred_test.extend(np.argmax(preds, axis=1))

y_true_test = np.array(y_true_test, dtype=np.int64)
y_pred_test = np.array(y_pred_test, dtype=np.int64)

acc = float(np.mean(y_true_test == y_pred_test))
print(f"\nOverall accuracy: {acc*100:.2f}%")

print("Test Classification Report:")
print(classification_report(y_true_test, y_pred_test, target_names=class_names, zero_division=0))

cm_test = confusion_matrix(y_true_test, y_pred_test, labels=range(len(class_names)))
plt.figure(figsize=(8, 6))
sns.heatmap(cm_test, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"{CROP_NAME} - Test Confusion Matrix")
plt.show()

# EXPORT TFLITE wrapper for Grad-CAM (two outputs)
# 1) Find last convolutional layer
def find_last_conv_layer(model: tf.keras.Model):
    for layer in reversed(model.layers):
        lname = layer.__class__.__name__.lower()
        if ("conv" in lname) or ("depthwiseconv" in lname) or ("separableconv" in lname):
            return layer.name
    raise ValueError("No convolutional layer found in the model. Can't create CAM without conv layers.")

last_conv_name = find_last_conv_layer(model)
print("Last conv layer:", last_conv_name)

# 2) Build wrapper model (preds, feature_maps)
last_conv_layer = model.get_layer(last_conv_name)
wrapper_model = tf.keras.Model(inputs=model.input, outputs=[model.output, last_conv_layer.output],
                               name=f"{model.name}_with_featuremaps")
wrapper_model.summary(line_length=120)

# 3) Save classifier Dense weights (for CAM on-device)
# Find dense layers in the head (expect two: penultimate and classifier)
dense_layers = [l for l in model.layers if isinstance(l, tf.keras.layers.Dense)]
if len(dense_layers) < 1:
    raise ValueError("No Dense layer found in the model head.")

# if there are >=2, last is classifier, second-last is penultimate
classifier_layer = dense_layers[-1]
W2, b2 = classifier_layer.get_weights()   # shape (units_prev, num_classes), (num_classes,)
if len(dense_layers) >= 2:
    penult_layer = dense_layers[-2]
    W1, b1 = penult_layer.get_weights()   # shape (channels, penult_units), (penult_units,)
    # Combined mapping from GAP channels -> classes
    # W_combined: (channels, num_classes)
    W_combined = np.matmul(W1, W2)        # (channels, num_classes)
    # Combined bias
    b_combined = np.matmul(b1, W2) + b2   # (num_classes,)
else:
    # If single Dense directly after GAP -> no penultimate; W2 is already channels->classes
    W_combined, b_combined = W2, b2

# Save combined weights which are compatible with feature maps (channels x num_classes)
np.save(OUT_DIR / f"{CROP_NAME.lower()}_classifier_weights_combined_{PRE_TRAINED_MODEL}.npy", W_combined)
np.save(OUT_DIR / f"{CROP_NAME.lower()}_classifier_bias_combined_{PRE_TRAINED_MODEL}.npy", b_combined)

print("Saved combined classifier weights:", W_combined.shape, " bias:", b_combined.shape)


# 4) Convert wrapper to TFLite; try normal, fallback to SELECT_TF_OPS if needed
TFLITE_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
converter = tf.lite.TFLiteConverter.from_keras_model(wrapper_model)

# try conversion; if converter needs TF ops, convert with SELECT_TF_OPS
try:
    tflite_model = converter.convert()
    TFLITE_PATH.write_bytes(tflite_model)
    print(f"✅ TFLite (two-output) saved to: {TFLITE_PATH}")
except Exception as e:
    print("⚠️ First conversion failed — attempting with SELECT_TF_OPS. Error:", e)
    converter.target_spec.supported_ops = [
        tf.lite.OpsSet.TFLITE_BUILTINS,
        tf.lite.OpsSet.SELECT_TF_OPS
    ]
    tflite_model = converter.convert()
    TFLITE_PATH.write_bytes(tflite_model)
    print(f"✅ TFLite (two-output, SELECT_TF_OPS) saved to: {TFLITE_PATH}")

print("Done. Files in:", OUT_DIR.resolve())
```

#### Model Evaluaton - Overall Accuracy

```python
import tensorflow as tf
import numpy as np
import cv2
import os
from pathlib import Path
from tqdm import tqdm
import pandas as pd

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_recall_fscore_support,
    matthews_corrcoef,
)

# ==============================
# CONFIG (edit these)
# ==============================
CROP_NAME = "Grape"
PRE_TRAINED_MODEL = "MobileNetV3Large_TrainTestVal"
# OUT_DIR = Path("/exported_models")
OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")
# SAVE_DIR = Path("/exported_models")
SAVE_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/Evaluation")
# SAVE_DIR = Path("../Datasets/Evaluation")
OUT_DIR.mkdir(parents=True, exist_ok=True)
SAVE_DIR.mkdir(parents=True, exist_ok=True)
DATASET_PATH = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated") / CROP_NAME
# DATASET_PATH = Path("../Datasets/plant-village-dataset-updated-aggregated") / CROP_NAME
MODEL_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
LABELS_PATH = Path(f"{CROP_NAME.lower()}_disease_labels.txt")
IMG_SIZE = (224, 224)
SCALE_0_1 = False  # set True only if model expects [0,1] inputs

# ==============================
# LOAD LABELS
# ==============================
assert LABELS_PATH.exists(), f"Labels file not found: {LABELS_PATH}"
with open(LABELS_PATH, "r") as f:
    class_names = [line.strip() for line in f if line.strip()]

num_classes = len(class_names)
print(f"✅ Loaded {num_classes} classes from {LABELS_PATH.name}:")
print(class_names)

# ==============================
# LOAD TFLITE MODEL
# ==============================
assert MODEL_PATH.exists(), f"TFLite model not found: {MODEL_PATH}"
interpreter = tf.lite.Interpreter(model_path=str(MODEL_PATH))
interpreter.allocate_tensors()
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# ==============================
# HELPERS
# ==============================
def preprocess_image(img_path: Path):
    """Read image, convert to RGB, resize and optionally scale."""
    img = cv2.imread(str(img_path))
    if img is None:
        raise ValueError(f"Failed to read image: {img_path}")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, IMG_SIZE)
    img = img.astype(np.float32)
    if SCALE_0_1:
        img = img / 255.0
    img = np.expand_dims(img, axis=0)  # [1, H, W, 3]
    return img

def predict_idx(img_path: Path) -> int:
    """Predict class index for a single image using the TFLite interpreter."""
    x = preprocess_image(img_path)
    interpreter.set_tensor(input_details[0]['index'], x)
    interpreter.invoke()
    out = interpreter.get_tensor(output_details[-1]['index'])[0]  # robust to multi-output tflite
    out = np.asarray(out).astype(np.float32)
    return int(np.argmax(out))

# ==============================
# COLLECT FILES + PER-FOLDER COUNTS
# ==============================
all_files = []
per_folder_counts = {}
label_to_index = {name: i for i, name in enumerate(class_names)}

for disease_name in class_names:
    folder = DATASET_PATH / disease_name
    if not folder.exists():
        print(f"⚠️ Skipping missing folder (in dataset): {folder}")
        per_folder_counts[disease_name] = 0
        continue
    count = 0
    for root, _, files in os.walk(folder):
        for f in files:
            if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.webp', '.jfif', '.tif', '.tiff')):
                all_files.append((Path(root) / f, label_to_index[disease_name]))
                count += 1
    per_folder_counts[disease_name] = count
    print(f"📁 Folder '{disease_name}': {count} images")

print(f"\n🔍 Found {len(all_files)} images for evaluation in total.")

# ==============================
# EVALUATE WITH PROGRESS BAR
# ==============================
y_true, y_pred = [], []
for img_path, true_idx in tqdm(all_files, desc=f"Evaluating {CROP_NAME}", ncols=100):
    pred_idx = predict_idx(img_path)
    y_true.append(true_idx)
    y_pred.append(pred_idx)

y_true = np.array(y_true, dtype=np.int32)
y_pred = np.array(y_pred, dtype=np.int32)

# ==============================
# CONFUSION MATRIX (print first as requested)
# ==============================
print("\n🧮 Confusion Matrix (rows=True, cols=Pred):")
cm = confusion_matrix(y_true, y_pred, labels=list(range(num_classes)))
print(cm)

# ==============================
# classification_report dict & TP/FP/FN/TN
# ==============================
report_dict = classification_report(y_true, y_pred, target_names=class_names, output_dict=True, zero_division=0)

TP = cm.diagonal().astype(np.float64)
FP = cm.sum(axis=0) - TP
FN = cm.sum(axis=1) - TP
TN = cm.sum() - (TP + FP + FN)

eps = 1e-12  # avoid division by zero

# Build per-class rows by combining sklearn report + manual fields
rows = []
supports = []
specificities = []
gdrs = []

for i, name in enumerate(class_names):
    prec = float(report_dict[name]["precision"])
    rec = float(report_dict[name]["recall"])   # sensitivity
    f1 = float(report_dict[name]["f1-score"])
    support = int(report_dict[name]["support"])
    supports.append(support)

    spec = float(TN[i] / (TN[i] + FP[i] + eps))
    gdr_val = float(np.sqrt(rec * spec))
    gdrs.append(gdr_val)
    specificities.append(spec)

    per_class_acc = float((TP[i] + TN[i]) / (TP[i] + TN[i] + FP[i] + FN[i] + eps))

    rows.append({
        "class": name,
        "support": support,
        "TP": int(TP[i]),
        "FP": int(FP[i]),
        "FN": int(FN[i]),
        "TN": int(TN[i]),
        "precision": prec,
        "recall (sensitivity)": rec,
        "specificity": spec,
        "f1": f1,
        "gdr (g-mean)": gdr_val,
        "per-class accuracy": per_class_acc,
    })

metrics_df = pd.DataFrame(rows)

# Round float columns to 4 decimal places for compact display and CSV
float_cols = ["precision", "recall (sensitivity)", "specificity", "f1", "gdr (g-mean)", "per-class accuracy"]
metrics_df[float_cols] = metrics_df[float_cols].round(4)

# Reorder columns for compactness
metrics_df = metrics_df[
    ["class", "support", "TP", "FP", "FN", "TN",
     "precision", "recall (sensitivity)", "specificity",
     "f1", "gdr (g-mean)", "per-class accuracy"]
]

# ==============================
# PRINT per-class table compactly (rounded)
# ==============================
print("\n📋 Per-class metrics (combined from classification_report + extras):")
# Use to_string with float_format to ensure 4 decimal places in console
print(metrics_df.to_string(index=False, float_format="{:.4f}".format))

# ==============================
# Macro & Weighted averages
# - precision/recall/f1 from classification_report dict (if present)
# - specificity/gdr: compute macro (mean) and weighted (support-weighted)
# ==============================
macro = report_dict.get("macro avg", {})
weighted = report_dict.get("weighted avg", {})

supports_arr = np.array(supports, dtype=np.float64)
macro_specificity = float(np.mean(specificities))
macro_gdr = float(np.mean(gdrs))
weighted_specificity = float(np.average(specificities, weights=supports_arr))
weighted_gdr = float(np.average(gdrs, weights=supports_arr))

# Safely get macro/weighted precision/recall/f1; fallback to sklearn calc if missing
if macro:
    macro_prec = macro.get("precision", np.nan)
    macro_rec = macro.get("recall", np.nan)
    macro_f1 = macro.get("f1-score", np.nan)
else:
    macro_prec, macro_rec, macro_f1, _ = precision_recall_fscore_support(y_true, y_pred, average='macro', zero_division=0)

if weighted:
    weighted_prec = weighted.get("precision", np.nan)
    weighted_rec = weighted.get("recall", np.nan)
    weighted_f1 = weighted.get("f1-score", np.nan)
else:
    weighted_prec, weighted_rec, weighted_f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=0)

# Round summary values to 4 decimals
def r(x): return float(np.round(x, 4)) if not (x is None or (isinstance(x, float) and np.isnan(x))) else np.nan

macro_prec, macro_rec, macro_f1 = r(macro_prec), r(macro_rec), r(macro_f1)
weighted_prec, weighted_rec, weighted_f1 = r(weighted_prec), r(weighted_rec), r(weighted_f1)
macro_specificity, weighted_specificity = r(macro_specificity), r(weighted_specificity)
macro_gdr, weighted_gdr = r(macro_gdr), r(weighted_gdr)

# ==============================
# Remaining overall metrics and save CSV
# ==============================
acc = accuracy_score(y_true, y_pred)
mcc = matthews_corrcoef(y_true, y_pred)

print("\n🔢 Overall metrics:")
print(f"Overall accuracy: {acc:.4f}")
print(f"MCC (Matthews Correlation Coefficient): {mcc:.4f}")

# Save rounded per-class CSV for thesis tables/plots
csv_out = SAVE_DIR / f"{CROP_NAME.lower()}_evaluation_metrics_{PRE_TRAINED_MODEL}.csv"
metrics_df.to_csv(csv_out, index=False, float_format="%.4f")
np.save(SAVE_DIR / f"{CROP_NAME.lower()}_confusion_matrix_{PRE_TRAINED_MODEL}.npy", cm)

print(f"\nSaved per-class metrics CSV to: {csv_out}")
print(f"Saved confusion matrix (.npy) to: {SAVE_DIR / f'{CROP_NAME.lower()}_confusion_matrix_{PRE_TRAINED_MODEL}.npy'}")

# Final thesis-ready summary (explicit precision and recall included)
print("\n=== Thesis-ready summary ===")
print(f"Overall accuracy: {acc:.4f}")
print(f"Macro  - Precision: {macro_prec:.4f} | Recall: {macro_rec:.4f} | F1: {macro_f1:.4f} | Specificity: {macro_specificity:.4f} | GDR (g-mean): {macro_gdr:.4f}")
print(f"Weighted - Precision: {weighted_prec:.4f} | Recall: {weighted_rec:.4f} | F1: {weighted_f1:.4f} | Specificity: {weighted_specificity:.4f} | GDR (g-mean): {weighted_gdr:.4f}")
print(f"MCC: {mcc:.4f}")
print("Per-class metrics saved to CSV and printed above.")
```

```python
# Single notebook cell: Generate Grad-CAM images (5 per class) and display in grid (5 per row)
# -----------------------------------------------------------------------------
import tensorflow as tf
import numpy as np
import cv2
import os
import random
from pathlib import Path
from tqdm import tqdm
import matplotlib.pyplot as plt
import math

# ---------------------- USER CONFIG ----------------------
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

CROP_NAME = "Grape"   # e.g. "Apple", "Tomato"
PRE_TRAINED_MODEL = "MobileNetV3Large_TrainTestVal"
OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")   # where tflite + weights .npy live and where outputs will be written
# Local
# DATASET_ROOT = Path("../Datasets/plant-village-dataset-updated-aggregated")
# Colab
DATASET_ROOT = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated")
IMG_SIZE = (224, 224)   # model input size
IMAGES_PER_CLASS = 5    # how many random images per class to create heatmaps for
SCALE_0_1 = False       # set True if model expects inputs scaled to [0,1]; False to use 0..255 floats
VERBOSE = True
# ---------------------------------------------------------

# Derived paths
DATASET_PATH = DATASET_ROOT / CROP_NAME
TFLITE_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
WEIGHTS_PATH = OUT_DIR / f"{CROP_NAME.lower()}_classifier_weights_combined_{PRE_TRAINED_MODEL}.npy"
BIAS_PATH = OUT_DIR / f"{CROP_NAME.lower()}_classifier_bias_combined_{PRE_TRAINED_MODEL}.npy"
LABELS_PATH = Path(f"{CROP_NAME.lower()}_disease_labels.txt")
SAVE_DIR = OUT_DIR / "gradcam" / CROP_NAME
SAVE_DIR.mkdir(parents=True, exist_ok=True)

# Sanity checks
assert DATASET_PATH.exists(), f"Dataset folder not found: {DATASET_PATH}"
assert TFLITE_PATH.exists(), f"TFLite file not found: {TFLITE_PATH}"
assert WEIGHTS_PATH.exists(), f"Weights file not found: {WEIGHTS_PATH}"
assert LABELS_PATH.exists(), f"Labels file not found: {LABELS_PATH}"

# Load labels
with open(LABELS_PATH, "r") as f:
    class_names = [ln.strip() for ln in f if ln.strip()]
num_classes = len(class_names)
if VERBOSE:
    print(f"Loaded {num_classes} labels: {class_names}")

# Load classifier weights & bias
W = np.load(WEIGHTS_PATH)   # expected shape (channels, num_classes) if GAP -> Dense
b = None
if BIAS_PATH.exists():
    b = np.load(BIAS_PATH)

if VERBOSE:
    print("Classifier weight shape:", W.shape, " bias shape:", None if b is None else b.shape)

# Load TFLite interpreter
interpreter = tf.lite.Interpreter(model_path=str(TFLITE_PATH))
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
if VERBOSE:
    print("TFLite Input details:", input_details)
    print("TFLite Output details:", output_details)

# Identify which output index is feature maps and which is probabilities:
feat_idx = None
prob_idx = None
for od in output_details:
    shape = od["shape"]
    if len(shape) == 4:
        feat_idx = od["index"]
    elif len(shape) == 2:
        prob_idx = od["index"]

if feat_idx is None or prob_idx is None:
    raise RuntimeError("Couldn't identify feature-map and prob outputs from TFLite. Check model outputs.")

if VERBOSE:
    print("Using feature-output index:", feat_idx, "prob-output index:", prob_idx)

# Helper: image preproc (assumes model contains preprocess step, so pass raw 0..255 floats)
def preprocess_image(path: Path):
    img_bgr = cv2.imread(str(path))
    if img_bgr is None:
        raise ValueError(f"Failed to load image {path}")
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    resized = cv2.resize(img_rgb, IMG_SIZE)
    arr = resized.astype(np.float32)
    if SCALE_0_1:
        arr = arr / 255.0
    return arr, resized  # arr used as model input, resized is used for overlay visualization (RGB uint8)

# Helper: compute grad-cam heatmap from feature maps and classifier weights
def compute_gradcam_from_featuremaps(feature_maps: np.ndarray, class_id: int, weights: np.ndarray):
    """
    feature_maps: (1, H, W, C) float32
    weights: (C, num_classes) or (num_classes, C)
    returns heatmap HxW normalized [0,1]
    """
    fm = feature_maps[0]   # H x W x C
    # find correct weight vector for class
    if weights.shape[0] == fm.shape[2] and weights.shape[1] == num_classes:
        w_k = weights[:, class_id]    # (C,)
    elif weights.shape[1] == fm.shape[2] and weights.shape[0] == num_classes:
        w_k = weights[class_id, :]
    else:
        w_try = weights.T
        if w_try.shape[0] == fm.shape[2]:
            w_k = w_try[:, class_id]
        else:
            raise ValueError(f"Classifier weights shape {weights.shape} incompatible with feature channels {fm.shape[2]}")

    heatmap = np.tensordot(fm, w_k, axes=([2], [0]))   # shape (H, W)
    heatmap = np.maximum(heatmap, 0)
    if np.max(heatmap) > 0:
        heatmap = heatmap / np.max(heatmap)
    else:
        heatmap = np.zeros_like(heatmap)
    return heatmap.astype(np.float32)

# Helper: overlay heatmap on image
def overlay_heatmap_on_image(orig_rgb, heatmap, alpha=0.45, colormap=cv2.COLORMAP_JET):
    hm_u8 = np.uint8(255 * heatmap)
    hm_color = cv2.applyColorMap(hm_u8, colormap)         # BGR
    hm_color = cv2.cvtColor(hm_color, cv2.COLOR_BGR2RGB)  # to RGB
    hm_resized = cv2.resize(hm_color, (orig_rgb.shape[1], orig_rgb.shape[0]))
    overlay = (0.5 * orig_rgb + alpha * hm_resized).astype(np.uint8)
    return overlay

# Collect up to IMAGES_PER_CLASS random images per class
samples = []
for cls in class_names:
    folder = DATASET_PATH / cls
    if not folder.exists():
        print(f"⚠️ Missing folder {folder}, skipping")
        continue
    files = [Path(folder) / f for f in os.listdir(folder) if f.lower().endswith(('.jpg','.jpeg','.png','.bmp','.tif','.tiff'))]
    if len(files) == 0:
        continue
    chosen = random.sample(files, k=min(IMAGES_PER_CLASS, len(files)))
    for p in chosen:
        samples.append((cls, p))

if VERBOSE:
    print(f"Selected {len(samples)} total images ({IMAGES_PER_CLASS} per class where available).")

# Run tflite inference, compute heatmaps, save overlays
records = []
for cls, img_path in tqdm(samples, desc="Generating Grad-CAM", ncols=100):
    img_arr, resized_rgb = preprocess_image(img_path)
    input_tensor = np.expand_dims(img_arr, axis=0).astype(np.float32)

    interpreter.set_tensor(input_details[0]['index'], input_tensor)
    interpreter.invoke()
    feat = interpreter.get_tensor(feat_idx)   # (1,Hf,Wf,C)
    probs = interpreter.get_tensor(prob_idx)  # (1,num_classes)
    probs = np.asarray(probs).astype(np.float32)[0]
    pred_idx = int(np.argmax(probs))
    pred_label = class_names[pred_idx] if pred_idx < len(class_names) else str(pred_idx)
    confidence = float(probs[pred_idx])

    heatmap = compute_gradcam_from_featuremaps(feat, pred_idx, W)
    overlay = overlay_heatmap_on_image(resized_rgb, heatmap, alpha=0.5)

    class_folder = SAVE_DIR / cls
    class_folder.mkdir(parents=True, exist_ok=True)

    base_name = img_path.stem
    out_heatmap_path = class_folder / f"{base_name}_heatmap.png"
    out_overlay_path = class_folder / f"{base_name}_overlay.png"
    out_orig_path = class_folder / f"{base_name}_orig.png"

    cv2.imwrite(str(out_heatmap_path), (255 * cv2.resize(heatmap, (IMG_SIZE[0], IMG_SIZE[1]))).astype(np.uint8))
    cv2.imwrite(str(out_overlay_path), cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))
    cv2.imwrite(str(out_orig_path), cv2.cvtColor(resized_rgb, cv2.COLOR_RGB2BGR))

    records.append({
        "true_class": cls,
        "image_path": str(img_path),
        "predicted_label": pred_label,
        "confidence": confidence,
        "overlay": str(out_overlay_path),
        "orig": str(out_orig_path),
        "heatmap": str(out_heatmap_path)
    })

print(f"\nSaved {len(records)} Grad-CAM images to: {SAVE_DIR}")

# ------------------ DISPLAY: grid with 5 images per row ------------------
# ------------------ DISPLAY: grid with 5 images per row, no horizontal scroll, dark bold titles ------------------
if len(records) == 0:
    print("No records to display.")
else:
    per_row = 5                       # show exactly 5 images per row
    n = len(records)
    rows = math.ceil(n / per_row)

    # Prevent horizontal scrolling by capping figure width (in inches).
    # Increase/decrease max_width_inch if you know your display width.
    max_width_inch = 15
    cell_size_inch = 2.8              # size per image cell (adjust if you want larger/smaller)
    fig_w = min(per_row * cell_size_inch, max_width_inch)
    fig_h = rows * cell_size_inch

    fig, axes = plt.subplots(rows, per_row, figsize=(fig_w, fig_h))
    # ensure axes is 2D array even if rows==1
    if rows == 1:
        axes = np.expand_dims(axes, axis=0)
    axes = np.array(axes).reshape(rows, per_row)

    for idx, rec in enumerate(records):
        r = idx // per_row
        c = idx % per_row
        ax = axes[r, c]
        img_bgr = cv2.imread(rec["overlay"])
        if img_bgr is None:
            img_bgr = cv2.imread(rec["orig"])  # fallback
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
        ax.imshow(img_rgb)
        title = f"T:{rec['true_class']}  P:{rec['predicted_label']} ({rec['confidence']*100:.1f}%)"
        ax.set_title(title, fontsize=9, color='black', fontweight='bold')   # dark bold font
        ax.axis("off")

    # hide unused subplots
    for idx in range(n, rows * per_row):
        r = idx // per_row
        c = idx % per_row
        axes[r, c].axis("off")

    plt.tight_layout()
    plt.show()
```

```python
import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.metrics import classification_report, confusion_matrix
import os
import shutil
from collections import defaultdict
import math
from collections import Counter
import pandas as pd

# Random seed setup
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

# Dataset path local/colab
# Colab
DATASET_PATH = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated")
# Local
# DATASET_PATH = Path("../Datasets/plant-village-dataset-updated-aggregated")

# Change crop name here
CROP_NAME = "Grape"
CROP_PATH = DATASET_PATH / CROP_NAME
IMG_SIZE = (224, 224)
# CPU
# BATCH_SIZE = 32
# GPU
BATCH_SIZE = 64
EPOCHS = 30
PRE_TRAINED_MODEL = "EfficientNetV2B1_TrainTestVal"
DROPOUT = 0.3
LR_RATE = 1e-4

OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")
OUT_DIR.mkdir(parents=True, exist_ok=True)


# Data Loader (70% / 15% / 15%)
# Count per class in each split
def count_per_class(file_paths, class_names):
    labels = [Path(f).parent.name for f in file_paths]
    counts = Counter(labels)
    return {cls: counts.get(cls, 0) for cls in class_names}

# Gather file list to compute exact counts
all_files = []
for root, _, files in os.walk(CROP_PATH):
    for f in files:
        if f.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp")):
            all_files.append(Path(root) / f)
N = len(all_files)
if N == 0:
    raise RuntimeError(f"No images found under {CROP_PATH}")

train_count = int(N * 0.70)
val_count = int(N * 0.15)
# Validate sum
test_count = N - train_count - val_count

print(f"Total images: {N} -> train: {train_count}, val: {val_count}, test: {test_count}")

# Use image_dataset_from_directory with validation_split=0.3:
rest_split = 0.3

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    CROP_PATH,
    validation_split=rest_split,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True
)

rest_ds = tf.keras.preprocessing.image_dataset_from_directory(
    CROP_PATH,
    validation_split=rest_split,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,   # batched
    shuffle=True
)

class_names = train_ds.class_names
num_classes = len(class_names)

# Print the count of images per disease for Train, Test, and Val
train_files = train_ds.file_paths
rest_files = [p for p in rest_ds.file_paths]
val_files = rest_files[:val_count]
test_files = rest_files[val_count:val_count+test_count]

train_counts = count_per_class(train_files, class_names)
val_counts = count_per_class(val_files, class_names)
test_counts = count_per_class(test_files, class_names)

# Create DataFrame for thesis table
df = pd.DataFrame({
    "Train": train_counts,
    "Val": val_counts,
    "Test": test_counts
})
print("\n Per-class image distribution:\n")
print(df)
column_sums = df.sum()

print("\n--------------------")
print("Sum of each column:")
print(column_sums)

# Flatten rest_ds and split deterministically by counts
rest_ds = rest_ds.unbatch()
val_ds = rest_ds.take(val_count)
test_ds = rest_ds.skip(val_count).take(test_count)

# Rebatch val & test for training pipeline; we'll also create finite eval datasets below
val_ds = val_ds.batch(BATCH_SIZE)
test_ds = test_ds.batch(BATCH_SIZE)

# Derive class_names from train_ds
print("\n--------------------")
print(f"✅ Training {CROP_NAME} model with {num_classes} classes")
print("Classes:", class_names)
train_ds_len = tf.data.experimental.cardinality(train_ds)
print(f"Total images: {N} -> train data per batch: {train_ds_len.numpy()}")

# Save class labels
with open(f"{CROP_NAME.lower()}_disease_labels.txt", "w") as f:
    for name in class_names:
        f.write(name + "\n")

# Data augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
    tf.keras.layers.RandomContrast(0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
    tf.keras.layers.GaussianNoise(0.05),
    tf.keras.layers.RandomBrightness(factor=0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
])

AUTOTUNE = tf.data.AUTOTUNE


# Pipelines

# TRAINING pipeline: cache -> shuffle -> augment -> repeat -> prefetch
train_ds = (
    train_ds
    .cache()
    .shuffle(1000)
    .map(lambda x, y: (data_augmentation(x, training=True), y),
         num_parallel_calls=AUTOTUNE)
    .repeat()   # <-- ADD repeat here so fit() can run many epochs
    .prefetch(buffer_size=AUTOTUNE)
)

# VALIDATION for training (use a repeated version for validation during fit)
val_ds_for_fit = (
    val_ds
    .cache()
    .repeat()
    .prefetch(buffer_size=AUTOTUNE)
)

# VAL pipeline used for final evaluation remains finite (val_ds)
# TEST pipeline remains finite
test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE)

# compute steps for training loops
steps_per_epoch = max(1, train_count // BATCH_SIZE)
validation_steps = max(1, val_count // BATCH_SIZE)
print(f"steps_per_epoch={steps_per_epoch}, validation_steps={validation_steps}")

# Model Building
base_model = tf.keras.applications.EfficientNetV2B1(
    input_shape=IMG_SIZE + (3,),
    include_top=False,
    weights="imagenet",
    include_preprocessing=True
)
base_model.trainable = False  # freeze backbone initially

# Input -> backbone -> head (we reuse base_model.input/output)
# x = base_model.output
# x = tf.keras.layers.GlobalAveragePooling2D()(x)
# x = tf.keras.layers.Dropout(0.3)(x)
# outputs = tf.keras.layers.Dense(num_classes, activation="softmax", name="predictions")(x)

x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(256, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(LR_RATE))(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dropout(DROPOUT)(x)
outputs = tf.keras.layers.Dense(num_classes, activation="softmax", name="predictions")(x)

model = tf.keras.Model(base_model.input, outputs, name=f"{PRE_TRAINED_MODEL}_{CROP_NAME}")

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# Training
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss", patience=10, restore_best_weights=True
)

checkpoint = tf.keras.callbacks.ModelCheckpoint(
    f"best_{CROP_NAME.lower()}_disease_model_{PRE_TRAINED_MODEL}.keras",
    save_best_only=True,
    monitor="val_accuracy",
    mode="max"
)

history = model.fit(
    train_ds,
    validation_data=val_ds_for_fit,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# Fine-tuning: unfreeze some top layers of base_model
base_model.trainable = True
fine_tune_at = int(len(base_model.layers) * 0.75)
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_fine = model.fit(
    train_ds,
    validation_data=val_ds_for_fit,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# Evaluate on validation
results = model.evaluate(val_ds)
print("Metrics - ")
for name, value in zip(model.metrics_names, results):
    print(f"{name}: {value:.4f}")

# Plots: Accuracy & Loss
plt.figure(figsize=(10, 6))
plt.plot(history.history.get('accuracy', []) + history_fine.history.get('accuracy', []), label="Training Accuracy")
plt.plot(history.history.get('val_accuracy', []) + history_fine.history.get('val_accuracy', []), label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title(f"{CROP_NAME} - Training & Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(history.history.get('loss', []) + history_fine.history.get('loss', []), label="Training Loss")
plt.plot(history.history.get('val_loss', []) + history_fine.history.get('val_loss', []), label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title(f"{CROP_NAME} - Training & Validation Loss")
plt.legend()
plt.grid(True)
plt.show()

# Validation Evaluation: Classification Report & Confusion Matrix on VAL set
# ==============================
y_true = []
y_pred = []

for batch_images, batch_labels in val_ds:
    preds = model.predict(batch_images, verbose=0)
    y_true.extend(batch_labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

# Safety: ensure lengths align
if len(class_names) != (int(max(max(y_true, default=0), max(y_pred, default=0))) + 1):
    print("Warning: class_names length does not match labels. Recomputing class_names from folders.")
    class_names = sorted([d.name for d in CROP_PATH.iterdir() if d.is_dir()])
    num_classes = len(class_names)
    print("New class_names:", class_names)

y_true = np.array(y_true, dtype=np.int64)
y_pred = np.array(y_pred, dtype=np.int64)

acc = float(np.mean(y_true == y_pred))
print(f"\n Overall accuracy: {acc*100:.2f}%")

print("Validation Classification Report:")
print(classification_report(y_true, y_pred,labels=list(range(num_classes)), target_names=class_names, zero_division=0))

cm = confusion_matrix(y_true, y_pred, labels=range(len(class_names)))
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"{CROP_NAME} - Validation Confusion Matrix")
plt.show()

# Testing Evaluation: classification report & confusion matrix on TEST set
y_true_test = []
y_pred_test = []

for batch_images, batch_labels in test_ds:
    preds = model.predict(batch_images, verbose=0)
    y_true_test.extend(batch_labels.numpy())
    y_pred_test.extend(np.argmax(preds, axis=1))

y_true_test = np.array(y_true_test, dtype=np.int64)
y_pred_test = np.array(y_pred_test, dtype=np.int64)

acc = float(np.mean(y_true_test == y_pred_test))
print(f"\nOverall accuracy: {acc*100:.2f}%")

print("Test Classification Report:")
print(classification_report(y_true_test, y_pred_test, target_names=class_names, zero_division=0))

cm_test = confusion_matrix(y_true_test, y_pred_test, labels=range(len(class_names)))
plt.figure(figsize=(8, 6))
sns.heatmap(cm_test, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"{CROP_NAME} - Test Confusion Matrix")
plt.show()

# EXPORT TFLITE wrapper for Grad-CAM (two outputs)
# 1) Find last convolutional layer
def find_last_conv_layer(model: tf.keras.Model):
    for layer in reversed(model.layers):
        lname = layer.__class__.__name__.lower()
        if ("conv" in lname) or ("depthwiseconv" in lname) or ("separableconv" in lname):
            return layer.name
    raise ValueError("No convolutional layer found in the model. Can't create CAM without conv layers.")

last_conv_name = find_last_conv_layer(model)
print("Last conv layer:", last_conv_name)

# 2) Build wrapper model (preds, feature_maps)
last_conv_layer = model.get_layer(last_conv_name)
wrapper_model = tf.keras.Model(inputs=model.input, outputs=[model.output, last_conv_layer.output],
                               name=f"{model.name}_with_featuremaps")
wrapper_model.summary(line_length=120)

# 3) Save classifier Dense weights (for CAM on-device)
# Find dense layers in the head (expect two: penultimate and classifier)
dense_layers = [l for l in model.layers if isinstance(l, tf.keras.layers.Dense)]
if len(dense_layers) < 1:
    raise ValueError("No Dense layer found in the model head.")

# if there are >=2, last is classifier, second-last is penultimate
classifier_layer = dense_layers[-1]
W2, b2 = classifier_layer.get_weights()   # shape (units_prev, num_classes), (num_classes,)
if len(dense_layers) >= 2:
    penult_layer = dense_layers[-2]
    W1, b1 = penult_layer.get_weights()   # shape (channels, penult_units), (penult_units,)
    # Combined mapping from GAP channels -> classes
    # W_combined: (channels, num_classes)
    W_combined = np.matmul(W1, W2)        # (channels, num_classes)
    # Combined bias
    b_combined = np.matmul(b1, W2) + b2   # (num_classes,)
else:
    # If single Dense directly after GAP -> no penultimate; W2 is already channels->classes
    W_combined, b_combined = W2, b2

# Save combined weights which are compatible with feature maps (channels x num_classes)
np.save(OUT_DIR / f"{CROP_NAME.lower()}_classifier_weights_combined_{PRE_TRAINED_MODEL}.npy", W_combined)
np.save(OUT_DIR / f"{CROP_NAME.lower()}_classifier_bias_combined_{PRE_TRAINED_MODEL}.npy", b_combined)

print("Saved combined classifier weights:", W_combined.shape, " bias:", b_combined.shape)


# 4) Convert wrapper to TFLite; try normal, fallback to SELECT_TF_OPS if needed
TFLITE_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
converter = tf.lite.TFLiteConverter.from_keras_model(wrapper_model)

# try conversion; if converter needs TF ops, convert with SELECT_TF_OPS
try:
    tflite_model = converter.convert()
    TFLITE_PATH.write_bytes(tflite_model)
    print(f"✅ TFLite (two-output) saved to: {TFLITE_PATH}")
except Exception as e:
    print("⚠️ First conversion failed — attempting with SELECT_TF_OPS. Error:", e)
    converter.target_spec.supported_ops = [
        tf.lite.OpsSet.TFLITE_BUILTINS,
        tf.lite.OpsSet.SELECT_TF_OPS
    ]
    tflite_model = converter.convert()
    TFLITE_PATH.write_bytes(tflite_model)
    print(f"✅ TFLite (two-output, SELECT_TF_OPS) saved to: {TFLITE_PATH}")

print("Done. Files in:", OUT_DIR.resolve())
```

#### EfficientNetV2B1 - Model Evaluaton - Overall Accuracy

```python
import tensorflow as tf
import numpy as np
import cv2
import os
from pathlib import Path
from tqdm import tqdm
import pandas as pd

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_recall_fscore_support,
    matthews_corrcoef,
)

# ==============================
# CONFIG (edit these)
# ==============================
CROP_NAME = "Grape"
PRE_TRAINED_MODEL = "EfficientNetV2B1_TrainTestVal"
OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")
SAVE_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/Evaluation")
# SAVE_DIR = Path("../Datasets/Evaluation")
OUT_DIR.mkdir(parents=True, exist_ok=True)
SAVE_DIR.mkdir(parents=True, exist_ok=True)
DATASET_PATH = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated") / CROP_NAME
# DATASET_PATH = Path("../Datasets/plant-village-dataset-updated-aggregated") / CROP_NAME
MODEL_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
LABELS_PATH = Path(f"{CROP_NAME.lower()}_disease_labels.txt")
IMG_SIZE = (224, 224)
SCALE_0_1 = False  # set True only if model expects [0,1] inputs

# ==============================
# LOAD LABELS
# ==============================
assert LABELS_PATH.exists(), f"Labels file not found: {LABELS_PATH}"
with open(LABELS_PATH, "r") as f:
    class_names = [line.strip() for line in f if line.strip()]

num_classes = len(class_names)
print(f"✅ Loaded {num_classes} classes from {LABELS_PATH.name}:")
print(class_names)

# ==============================
# LOAD TFLITE MODEL
# ==============================
assert MODEL_PATH.exists(), f"TFLite model not found: {MODEL_PATH}"
interpreter = tf.lite.Interpreter(model_path=str(MODEL_PATH))
interpreter.allocate_tensors()
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# ==============================
# HELPERS
# ==============================
def preprocess_image(img_path: Path):
    """Read image, convert to RGB, resize and optionally scale."""
    img = cv2.imread(str(img_path))
    if img is None:
        raise ValueError(f"Failed to read image: {img_path}")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, IMG_SIZE)
    img = img.astype(np.float32)
    if SCALE_0_1:
        img = img / 255.0
    img = np.expand_dims(img, axis=0)  # [1, H, W, 3]
    return img

def predict_idx(img_path: Path) -> int:
    """Predict class index for a single image using the TFLite interpreter."""
    x = preprocess_image(img_path)
    interpreter.set_tensor(input_details[0]['index'], x)
    interpreter.invoke()
    out = interpreter.get_tensor(output_details[-1]['index'])[0]  # robust to multi-output tflite
    out = np.asarray(out).astype(np.float32)
    return int(np.argmax(out))

# ==============================
# COLLECT FILES + PER-FOLDER COUNTS
# ==============================
all_files = []
per_folder_counts = {}
label_to_index = {name: i for i, name in enumerate(class_names)}

for disease_name in class_names:
    folder = DATASET_PATH / disease_name
    if not folder.exists():
        print(f"⚠️ Skipping missing folder (in dataset): {folder}")
        per_folder_counts[disease_name] = 0
        continue
    count = 0
    for root, _, files in os.walk(folder):
        for f in files:
            if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.webp', '.jfif', '.tif', '.tiff')):
                all_files.append((Path(root) / f, label_to_index[disease_name]))
                count += 1
    per_folder_counts[disease_name] = count
    print(f"📁 Folder '{disease_name}': {count} images")

print(f"\n🔍 Found {len(all_files)} images for evaluation in total.")

# ==============================
# EVALUATE WITH PROGRESS BAR
# ==============================
y_true, y_pred = [], []
for img_path, true_idx in tqdm(all_files, desc=f"Evaluating {CROP_NAME}", ncols=100):
    pred_idx = predict_idx(img_path)
    y_true.append(true_idx)
    y_pred.append(pred_idx)

y_true = np.array(y_true, dtype=np.int32)
y_pred = np.array(y_pred, dtype=np.int32)

# ==============================
# CONFUSION MATRIX (print first as requested)
# ==============================
print("\n🧮 Confusion Matrix (rows=True, cols=Pred):")
cm = confusion_matrix(y_true, y_pred, labels=list(range(num_classes)))
print(cm)

# ==============================
# classification_report dict & TP/FP/FN/TN
# ==============================
report_dict = classification_report(y_true, y_pred, target_names=class_names, output_dict=True, zero_division=0)

TP = cm.diagonal().astype(np.float64)
FP = cm.sum(axis=0) - TP
FN = cm.sum(axis=1) - TP
TN = cm.sum() - (TP + FP + FN)

eps = 1e-12  # avoid division by zero

# Build per-class rows by combining sklearn report + manual fields
rows = []
supports = []
specificities = []
gdrs = []

for i, name in enumerate(class_names):
    prec = float(report_dict[name]["precision"])
    rec = float(report_dict[name]["recall"])   # sensitivity
    f1 = float(report_dict[name]["f1-score"])
    support = int(report_dict[name]["support"])
    supports.append(support)

    spec = float(TN[i] / (TN[i] + FP[i] + eps))
    gdr_val = float(np.sqrt(rec * spec))
    gdrs.append(gdr_val)
    specificities.append(spec)

    per_class_acc = float((TP[i] + TN[i]) / (TP[i] + TN[i] + FP[i] + FN[i] + eps))

    rows.append({
        "class": name,
        "support": support,
        "TP": int(TP[i]),
        "FP": int(FP[i]),
        "FN": int(FN[i]),
        "TN": int(TN[i]),
        "precision": prec,
        "recall (sensitivity)": rec,
        "specificity": spec,
        "f1": f1,
        "gdr (g-mean)": gdr_val,
        "per-class accuracy": per_class_acc,
    })

metrics_df = pd.DataFrame(rows)

# Round float columns to 4 decimal places for compact display and CSV
float_cols = ["precision", "recall (sensitivity)", "specificity", "f1", "gdr (g-mean)", "per-class accuracy"]
metrics_df[float_cols] = metrics_df[float_cols].round(4)

# Reorder columns for compactness
metrics_df = metrics_df[
    ["class", "support", "TP", "FP", "FN", "TN",
     "precision", "recall (sensitivity)", "specificity",
     "f1", "gdr (g-mean)", "per-class accuracy"]
]

# ==============================
# PRINT per-class table compactly (rounded)
# ==============================
print("\n📋 Per-class metrics (combined from classification_report + extras):")
# Use to_string with float_format to ensure 4 decimal places in console
print(metrics_df.to_string(index=False, float_format="{:.4f}".format))

# ==============================
# Macro & Weighted averages
# - precision/recall/f1 from classification_report dict (if present)
# - specificity/gdr: compute macro (mean) and weighted (support-weighted)
# ==============================
macro = report_dict.get("macro avg", {})
weighted = report_dict.get("weighted avg", {})

supports_arr = np.array(supports, dtype=np.float64)
macro_specificity = float(np.mean(specificities))
macro_gdr = float(np.mean(gdrs))
weighted_specificity = float(np.average(specificities, weights=supports_arr))
weighted_gdr = float(np.average(gdrs, weights=supports_arr))

# Safely get macro/weighted precision/recall/f1; fallback to sklearn calc if missing
if macro:
    macro_prec = macro.get("precision", np.nan)
    macro_rec = macro.get("recall", np.nan)
    macro_f1 = macro.get("f1-score", np.nan)
else:
    macro_prec, macro_rec, macro_f1, _ = precision_recall_fscore_support(y_true, y_pred, average='macro', zero_division=0)

if weighted:
    weighted_prec = weighted.get("precision", np.nan)
    weighted_rec = weighted.get("recall", np.nan)
    weighted_f1 = weighted.get("f1-score", np.nan)
else:
    weighted_prec, weighted_rec, weighted_f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=0)

# Round summary values to 4 decimals
def r(x): return float(np.round(x, 4)) if not (x is None or (isinstance(x, float) and np.isnan(x))) else np.nan

macro_prec, macro_rec, macro_f1 = r(macro_prec), r(macro_rec), r(macro_f1)
weighted_prec, weighted_rec, weighted_f1 = r(weighted_prec), r(weighted_rec), r(weighted_f1)
macro_specificity, weighted_specificity = r(macro_specificity), r(weighted_specificity)
macro_gdr, weighted_gdr = r(macro_gdr), r(weighted_gdr)

# ==============================
# Remaining overall metrics and save CSV
# ==============================
acc = accuracy_score(y_true, y_pred)
mcc = matthews_corrcoef(y_true, y_pred)

print("\n🔢 Overall metrics:")
print(f"Overall accuracy: {acc:.4f}")
print(f"MCC (Matthews Correlation Coefficient): {mcc:.4f}")

# Save rounded per-class CSV for thesis tables/plots
csv_out = SAVE_DIR / f"{CROP_NAME.lower()}_evaluation_metrics_{PRE_TRAINED_MODEL}.csv"
metrics_df.to_csv(csv_out, index=False, float_format="%.4f")
np.save(SAVE_DIR / f"{CROP_NAME.lower()}_confusion_matrix_{PRE_TRAINED_MODEL}.npy", cm)

print(f"\nSaved per-class metrics CSV to: {csv_out}")
print(f"Saved confusion matrix (.npy) to: {SAVE_DIR / f'{CROP_NAME.lower()}_confusion_matrix_{PRE_TRAINED_MODEL}.npy'}")

# Final thesis-ready summary (explicit precision and recall included)
print("\n=== Thesis-ready summary ===")
print(f"Overall accuracy: {acc:.4f}")
print(f"Macro  - Precision: {macro_prec:.4f} | Recall: {macro_rec:.4f} | F1: {macro_f1:.4f} | Specificity: {macro_specificity:.4f} | GDR (g-mean): {macro_gdr:.4f}")
print(f"Weighted - Precision: {weighted_prec:.4f} | Recall: {weighted_rec:.4f} | F1: {weighted_f1:.4f} | Specificity: {weighted_specificity:.4f} | GDR (g-mean): {weighted_gdr:.4f}")
print(f"MCC: {mcc:.4f}")
print("Per-class metrics saved to CSV and printed above.")
```

#### EfficientNetV2B1 - Model Evaluation - GRAD CAM

```python
# Single notebook cell: Generate Grad-CAM images (5 per class) and display in grid (5 per row)
# -----------------------------------------------------------------------------
import tensorflow as tf
import numpy as np
import cv2
import os
import random
from pathlib import Path
from tqdm import tqdm
import matplotlib.pyplot as plt
import math

# ---------------------- USER CONFIG ----------------------
SEED = 42
random.seed(SEED)
np.random.seed(SEED)

CROP_NAME = "Grape"   # e.g. "Apple", "Tomato"
PRE_TRAINED_MODEL = "EfficientNetV2B1_TrainTestVal"
OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")   # where tflite + weights .npy live and where outputs will be written
# Local
# DATASET_ROOT = Path("../Datasets/plant-village-dataset-updated-aggregated")
# Colab
DATASET_ROOT = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated")
IMG_SIZE = (224, 224)   # model input size
IMAGES_PER_CLASS = 5    # how many random images per class to create heatmaps for
SCALE_0_1 = False       # set True if model expects inputs scaled to [0,1]; False to use 0..255 floats
VERBOSE = True
# ---------------------------------------------------------

# Derived paths
DATASET_PATH = DATASET_ROOT / CROP_NAME
TFLITE_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
WEIGHTS_PATH = OUT_DIR / f"{CROP_NAME.lower()}_classifier_weights_combined_{PRE_TRAINED_MODEL}.npy"
BIAS_PATH = OUT_DIR / f"{CROP_NAME.lower()}_classifier_bias_combined_{PRE_TRAINED_MODEL}.npy"
LABELS_PATH = Path(f"{CROP_NAME.lower()}_disease_labels.txt")
SAVE_DIR = OUT_DIR / "gradcam" / CROP_NAME
SAVE_DIR.mkdir(parents=True, exist_ok=True)

# Sanity checks
assert DATASET_PATH.exists(), f"Dataset folder not found: {DATASET_PATH}"
assert TFLITE_PATH.exists(), f"TFLite file not found: {TFLITE_PATH}"
assert WEIGHTS_PATH.exists(), f"Weights file not found: {WEIGHTS_PATH}"
assert LABELS_PATH.exists(), f"Labels file not found: {LABELS_PATH}"

# Load labels
with open(LABELS_PATH, "r") as f:
    class_names = [ln.strip() for ln in f if ln.strip()]
num_classes = len(class_names)
if VERBOSE:
    print(f"Loaded {num_classes} labels: {class_names}")

# Load classifier weights & bias
W = np.load(WEIGHTS_PATH)   # expected shape (channels, num_classes) if GAP -> Dense
b = None
if BIAS_PATH.exists():
    b = np.load(BIAS_PATH)

if VERBOSE:
    print("Classifier weight shape:", W.shape, " bias shape:", None if b is None else b.shape)

# Load TFLite interpreter
interpreter = tf.lite.Interpreter(model_path=str(TFLITE_PATH))
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
if VERBOSE:
    print("TFLite Input details:", input_details)
    print("TFLite Output details:", output_details)

# Identify which output index is feature maps and which is probabilities:
feat_idx = None
prob_idx = None
for od in output_details:
    shape = od["shape"]
    if len(shape) == 4:
        feat_idx = od["index"]
    elif len(shape) == 2:
        prob_idx = od["index"]

if feat_idx is None or prob_idx is None:
    raise RuntimeError("Couldn't identify feature-map and prob outputs from TFLite. Check model outputs.")

if VERBOSE:
    print("Using feature-output index:", feat_idx, "prob-output index:", prob_idx)

# Helper: image preproc (assumes model contains preprocess step, so pass raw 0..255 floats)
def preprocess_image(path: Path):
    img_bgr = cv2.imread(str(path))
    if img_bgr is None:
        raise ValueError(f"Failed to load image {path}")
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    resized = cv2.resize(img_rgb, IMG_SIZE)
    arr = resized.astype(np.float32)
    if SCALE_0_1:
        arr = arr / 255.0
    return arr, resized  # arr used as model input, resized is used for overlay visualization (RGB uint8)

# Helper: compute grad-cam heatmap from feature maps and classifier weights
def compute_gradcam_from_featuremaps(feature_maps: np.ndarray, class_id: int, weights: np.ndarray):
    """
    feature_maps: (1, H, W, C) float32
    weights: (C, num_classes) or (num_classes, C)
    returns heatmap HxW normalized [0,1]
    """
    fm = feature_maps[0]   # H x W x C
    # find correct weight vector for class
    if weights.shape[0] == fm.shape[2] and weights.shape[1] == num_classes:
        w_k = weights[:, class_id]    # (C,)
    elif weights.shape[1] == fm.shape[2] and weights.shape[0] == num_classes:
        w_k = weights[class_id, :]
    else:
        w_try = weights.T
        if w_try.shape[0] == fm.shape[2]:
            w_k = w_try[:, class_id]
        else:
            raise ValueError(f"Classifier weights shape {weights.shape} incompatible with feature channels {fm.shape[2]}")

    heatmap = np.tensordot(fm, w_k, axes=([2], [0]))   # shape (H, W)
    heatmap = np.maximum(heatmap, 0)
    if np.max(heatmap) > 0:
        heatmap = heatmap / np.max(heatmap)
    else:
        heatmap = np.zeros_like(heatmap)
    return heatmap.astype(np.float32)

# Helper: overlay heatmap on image
def overlay_heatmap_on_image(orig_rgb, heatmap, alpha=0.45, colormap=cv2.COLORMAP_JET):
    hm_u8 = np.uint8(255 * heatmap)
    hm_color = cv2.applyColorMap(hm_u8, colormap)         # BGR
    hm_color = cv2.cvtColor(hm_color, cv2.COLOR_BGR2RGB)  # to RGB
    hm_resized = cv2.resize(hm_color, (orig_rgb.shape[1], orig_rgb.shape[0]))
    overlay = (0.5 * orig_rgb + alpha * hm_resized).astype(np.uint8)
    return overlay

# Collect up to IMAGES_PER_CLASS random images per class
samples = []
for cls in class_names:
    folder = DATASET_PATH / cls
    if not folder.exists():
        print(f"⚠️ Missing folder {folder}, skipping")
        continue
    files = [Path(folder) / f for f in os.listdir(folder) if f.lower().endswith(('.jpg','.jpeg','.png','.bmp','.tif','.tiff'))]
    if len(files) == 0:
        continue
    chosen = random.sample(files, k=min(IMAGES_PER_CLASS, len(files)))
    for p in chosen:
        samples.append((cls, p))

if VERBOSE:
    print(f"Selected {len(samples)} total images ({IMAGES_PER_CLASS} per class where available).")

# Run tflite inference, compute heatmaps, save overlays
records = []
for cls, img_path in tqdm(samples, desc="Generating Grad-CAM", ncols=100):
    img_arr, resized_rgb = preprocess_image(img_path)
    input_tensor = np.expand_dims(img_arr, axis=0).astype(np.float32)

    interpreter.set_tensor(input_details[0]['index'], input_tensor)
    interpreter.invoke()
    feat = interpreter.get_tensor(feat_idx)   # (1,Hf,Wf,C)
    probs = interpreter.get_tensor(prob_idx)  # (1,num_classes)
    probs = np.asarray(probs).astype(np.float32)[0]
    pred_idx = int(np.argmax(probs))
    pred_label = class_names[pred_idx] if pred_idx < len(class_names) else str(pred_idx)
    confidence = float(probs[pred_idx])

    heatmap = compute_gradcam_from_featuremaps(feat, pred_idx, W)
    overlay = overlay_heatmap_on_image(resized_rgb, heatmap, alpha=0.5)

    class_folder = SAVE_DIR / cls
    class_folder.mkdir(parents=True, exist_ok=True)

    base_name = img_path.stem
    out_heatmap_path = class_folder / f"{base_name}_heatmap.png"
    out_overlay_path = class_folder / f"{base_name}_overlay.png"
    out_orig_path = class_folder / f"{base_name}_orig.png"

    cv2.imwrite(str(out_heatmap_path), (255 * cv2.resize(heatmap, (IMG_SIZE[0], IMG_SIZE[1]))).astype(np.uint8))
    cv2.imwrite(str(out_overlay_path), cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))
    cv2.imwrite(str(out_orig_path), cv2.cvtColor(resized_rgb, cv2.COLOR_RGB2BGR))

    records.append({
        "true_class": cls,
        "image_path": str(img_path),
        "predicted_label": pred_label,
        "confidence": confidence,
        "overlay": str(out_overlay_path),
        "orig": str(out_orig_path),
        "heatmap": str(out_heatmap_path)
    })

print(f"\nSaved {len(records)} Grad-CAM images to: {SAVE_DIR}")

# ------------------ DISPLAY: grid with 5 images per row ------------------
# ------------------ DISPLAY: grid with 5 images per row, no horizontal scroll, dark bold titles ------------------
if len(records) == 0:
    print("No records to display.")
else:
    per_row = 5                       # show exactly 5 images per row
    n = len(records)
    rows = math.ceil(n / per_row)

    # Prevent horizontal scrolling by capping figure width (in inches).
    # Increase/decrease max_width_inch if you know your display width.
    max_width_inch = 15
    cell_size_inch = 2.8              # size per image cell (adjust if you want larger/smaller)
    fig_w = min(per_row * cell_size_inch, max_width_inch)
    fig_h = rows * cell_size_inch

    fig, axes = plt.subplots(rows, per_row, figsize=(fig_w, fig_h))
    # ensure axes is 2D array even if rows==1
    if rows == 1:
        axes = np.expand_dims(axes, axis=0)
    axes = np.array(axes).reshape(rows, per_row)

    for idx, rec in enumerate(records):
        r = idx // per_row
        c = idx % per_row
        ax = axes[r, c]
        img_bgr = cv2.imread(rec["overlay"])
        if img_bgr is None:
            img_bgr = cv2.imread(rec["orig"])  # fallback
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
        ax.imshow(img_rgb)
        title = f"T:{rec['true_class']}  P:{rec['predicted_label']} ({rec['confidence']*100:.1f}%)"
        ax.set_title(title, fontsize=9, color='black', fontweight='bold')   # dark bold font
        ax.axis("off")

    # hide unused subplots
    for idx in range(n, rows * per_row):
        r = idx // per_row
        c = idx % per_row
        axes[r, c].axis("off")

    plt.tight_layout()
    plt.show()

```

```python
import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.metrics import classification_report, confusion_matrix
import os
import shutil
from collections import defaultdict
import math
from collections import Counter
import pandas as pd

# Random seed setup
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

# Dataset path local/colab
# Colab
DATASET_PATH = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated")
# Local
# DATASET_PATH = Path("../Datasets/plant-village-dataset-updated-aggregated")

# Change crop name here
CROP_NAME = "Grape"
CROP_PATH = DATASET_PATH / CROP_NAME
IMG_SIZE = (224, 224)
# CPU
# BATCH_SIZE = 32
# GPU
BATCH_SIZE = 64
EPOCHS = 30
PRE_TRAINED_MODEL = "Xception_TrainTestVal"
DROPOUT = 0.3
LR_RATE = 1e-4

OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# add this (minimal) — Xception preprocessing
preprocess_fn = tf.keras.applications.xception.preprocess_input


# Data Loader (70% / 15% / 15%)
# Count per class in each split
def count_per_class(file_paths, class_names):
    labels = [Path(f).parent.name for f in file_paths]
    counts = Counter(labels)
    return {cls: counts.get(cls, 0) for cls in class_names}

# Gather file list to compute exact counts
all_files = []
for root, _, files in os.walk(CROP_PATH):
    for f in files:
        if f.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp")):
            all_files.append(Path(root) / f)
N = len(all_files)
if N == 0:
    raise RuntimeError(f"No images found under {CROP_PATH}")

train_count = int(N * 0.70)
val_count = int(N * 0.15)
# Validate sum
test_count = N - train_count - val_count

print(f"Total images: {N} -> train: {train_count}, val: {val_count}, test: {test_count}")

# Use image_dataset_from_directory with validation_split=0.3:
rest_split = 0.3

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    CROP_PATH,
    validation_split=rest_split,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True
)

rest_ds = tf.keras.preprocessing.image_dataset_from_directory(
    CROP_PATH,
    validation_split=rest_split,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,   # batched
    shuffle=True
)

class_names = train_ds.class_names
num_classes = len(class_names)

# Print the count of images per disease for Train, Test, and Val
train_files = train_ds.file_paths
rest_files = [p for p in rest_ds.file_paths]
val_files = rest_files[:val_count]
test_files = rest_files[val_count:val_count+test_count]

train_counts = count_per_class(train_files, class_names)
val_counts = count_per_class(val_files, class_names)
test_counts = count_per_class(test_files, class_names)

# Create DataFrame for thesis table
df = pd.DataFrame({
    "Train": train_counts,
    "Val": val_counts,
    "Test": test_counts
})
print("\n Per-class image distribution:\n")
print(df)
column_sums = df.sum()

print("\n--------------------")
print("Sum of each column:")
print(column_sums)

# Flatten rest_ds and split deterministically by counts
rest_ds = rest_ds.unbatch()
val_ds = rest_ds.take(val_count)
test_ds = rest_ds.skip(val_count).take(test_count)

# Rebatch val & test for training pipeline; we'll also create finite eval datasets below
val_ds = val_ds.batch(BATCH_SIZE)
test_ds = test_ds.batch(BATCH_SIZE)

# Derive class_names from train_ds
print("\n--------------------")
print(f"✅ Training {CROP_NAME} model with {num_classes} classes")
print("Classes:", class_names)
train_ds_len = tf.data.experimental.cardinality(train_ds)
print(f"Total images: {N} -> train data per batch: {train_ds_len.numpy()}")

# Save class labels
with open(f"{CROP_NAME.lower()}_disease_labels.txt", "w") as f:
    for name in class_names:
        f.write(name + "\n")

# Data augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
    tf.keras.layers.RandomContrast(0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
    tf.keras.layers.GaussianNoise(0.05),
    tf.keras.layers.RandomBrightness(factor=0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
])

AUTOTUNE = tf.data.AUTOTUNE


# Pipelines

# TRAINING pipeline: cache -> shuffle -> augment -> preprocess -> repeat -> prefetch
train_ds = (
    train_ds
    .cache()
    .shuffle(1000)
    .map(lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=AUTOTUNE)
    .map(lambda x, y: (preprocess_fn(tf.cast(x, tf.float32)), y), num_parallel_calls=AUTOTUNE)   # <-- added preprocessing
    .repeat()   # <-- keep repeat so fit() can run many epochs
    .prefetch(buffer_size=AUTOTUNE)
)

# VALIDATION for training (use a repeated version for validation during fit)
val_ds_for_fit = (
    val_ds
    .cache()
    .map(lambda x, y: (preprocess_fn(tf.cast(x, tf.float32)), y), num_parallel_calls=AUTOTUNE)
    .repeat()
    .prefetch(buffer_size=AUTOTUNE)
)

val_ds_for_eval = (
    val_ds
    .cache()
    .map(lambda x, y: (preprocess_fn(tf.cast(x, tf.float32)), y), num_parallel_calls=AUTOTUNE)
    .prefetch(buffer_size=AUTOTUNE)
)

# VAL pipeline used for final evaluation remains finite (val_ds)
# TEST pipeline remains finite
test_ds = (
    test_ds
    .cache()
    .map(lambda x, y: (preprocess_fn(tf.cast(x, tf.float32)), y), num_parallel_calls=AUTOTUNE)
    .prefetch(buffer_size=AUTOTUNE)
)


# compute steps for training loops
steps_per_epoch = max(1, train_count // BATCH_SIZE)
validation_steps = max(1, val_count // BATCH_SIZE)
print(f"steps_per_epoch={steps_per_epoch}, validation_steps={validation_steps}")

# Model Building
base_model = tf.keras.applications.Xception(
    input_shape=IMG_SIZE + (3,),
    include_top=False,
    weights="imagenet",
    # include_preprocessing=True
)
base_model.trainable = False  # freeze backbone initially

# Input -> backbone -> head (we reuse base_model.input/output)
# x = base_model.output
# x = tf.keras.layers.GlobalAveragePooling2D()(x)
# x = tf.keras.layers.Dropout(0.3)(x)
# outputs = tf.keras.layers.Dense(num_classes, activation="softmax", name="predictions")(x)

x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(256, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(LR_RATE))(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dropout(DROPOUT)(x)
outputs = tf.keras.layers.Dense(num_classes, activation="softmax", name="predictions")(x)

model = tf.keras.Model(base_model.input, outputs, name=f"{PRE_TRAINED_MODEL}_{CROP_NAME}")

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# Training
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss", patience=7, restore_best_weights=True
)

checkpoint = tf.keras.callbacks.ModelCheckpoint(
    f"best_{CROP_NAME.lower()}_disease_model_{PRE_TRAINED_MODEL}.keras",
    save_best_only=True,
    monitor="val_accuracy",
    mode="max"
)

history = model.fit(
    train_ds,
    validation_data=val_ds_for_fit,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# Fine-tuning: unfreeze some top layers of base_model
base_model.trainable = True
fine_tune_at = int(len(base_model.layers) * 0.75)
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_fine = model.fit(
    train_ds,
    validation_data=val_ds_for_fit,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# Evaluate on validation
results = model.evaluate(val_ds_for_eval)
print("Metrics - ")
for name, value in zip(model.metrics_names, results):
    print(f"{name}: {value:.4f}")

# Plots: Accuracy & Loss
plt.figure(figsize=(10, 6))
plt.plot(history.history.get('accuracy', []) + history_fine.history.get('accuracy', []), label="Training Accuracy")
plt.plot(history.history.get('val_accuracy', []) + history_fine.history.get('val_accuracy', []), label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title(f"{CROP_NAME} - Training & Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(history.history.get('loss', []) + history_fine.history.get('loss', []), label="Training Loss")
plt.plot(history.history.get('val_loss', []) + history_fine.history.get('val_loss', []), label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title(f"{CROP_NAME} - Training & Validation Loss")
plt.legend()
plt.grid(True)
plt.show()

# Validation Evaluation: Classification Report & Confusion Matrix on VAL set
# ==============================
y_true = []
y_pred = []

for batch_images, batch_labels in val_ds_for_eval:
    preds = model.predict(batch_images, verbose=0)
    y_true.extend(batch_labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

# Safety: ensure lengths align
if len(class_names) != (int(max(max(y_true, default=0), max(y_pred, default=0))) + 1):
    print("Warning: class_names length does not match labels. Recomputing class_names from folders.")
    class_names = sorted([d.name for d in CROP_PATH.iterdir() if d.is_dir()])
    num_classes = len(class_names)
    print("New class_names:", class_names)

y_true = np.array(y_true, dtype=np.int64)
y_pred = np.array(y_pred, dtype=np.int64)

acc = float(np.mean(y_true == y_pred))
print(f"\n Overall accuracy: {acc*100:.2f}%")

print("Validation Classification Report:")
print(classification_report(y_true, y_pred,labels=list(range(num_classes)), target_names=class_names, zero_division=0))

cm = confusion_matrix(y_true, y_pred, labels=range(len(class_names)))
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"{CROP_NAME} - Validation Confusion Matrix")
plt.show()

# Testing Evaluation: classification report & confusion matrix on TEST set
y_true_test = []
y_pred_test = []

for batch_images, batch_labels in test_ds:
    preds = model.predict(batch_images, verbose=0)
    y_true_test.extend(batch_labels.numpy())
    y_pred_test.extend(np.argmax(preds, axis=1))

y_true_test = np.array(y_true_test, dtype=np.int64)
y_pred_test = np.array(y_pred_test, dtype=np.int64)

acc = float(np.mean(y_true_test == y_pred_test))
print(f"\nOverall accuracy: {acc*100:.2f}%")

print("Test Classification Report:")
print(classification_report(y_true_test, y_pred_test, target_names=class_names, zero_division=0))

cm_test = confusion_matrix(y_true_test, y_pred_test, labels=range(len(class_names)))
plt.figure(figsize=(8, 6))
sns.heatmap(cm_test, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"{CROP_NAME} - Test Confusion Matrix")
plt.show()

# EXPORT TFLITE wrapper for Grad-CAM (two outputs)
# 1) Find last convolutional layer
def find_last_conv_layer(model: tf.keras.Model):
    for layer in reversed(model.layers):
        lname = layer.__class__.__name__.lower()
        if ("conv" in lname) or ("depthwiseconv" in lname) or ("separableconv" in lname):
            return layer.name
    raise ValueError("No convolutional layer found in the model. Can't create CAM without conv layers.")

last_conv_name = find_last_conv_layer(model)
print("Last conv layer:", last_conv_name)

# 2) Build wrapper model (preds, feature_maps)
last_conv_layer = model.get_layer(last_conv_name)
wrapper_model = tf.keras.Model(inputs=model.input, outputs=[model.output, last_conv_layer.output],
                               name=f"{model.name}_with_featuremaps")
wrapper_model.summary(line_length=120)

# 3) Save classifier Dense weights (for CAM on-device)
# Find dense layers in the head (expect two: penultimate and classifier)
dense_layers = [l for l in model.layers if isinstance(l, tf.keras.layers.Dense)]
if len(dense_layers) < 1:
    raise ValueError("No Dense layer found in the model head.")

# if there are >=2, last is classifier, second-last is penultimate
classifier_layer = dense_layers[-1]
W2, b2 = classifier_layer.get_weights()   # shape (units_prev, num_classes), (num_classes,)
if len(dense_layers) >= 2:
    penult_layer = dense_layers[-2]
    W1, b1 = penult_layer.get_weights()   # shape (channels, penult_units), (penult_units,)
    # Combined mapping from GAP channels -> classes
    # W_combined: (channels, num_classes)
    W_combined = np.matmul(W1, W2)        # (channels, num_classes)
    # Combined bias
    b_combined = np.matmul(b1, W2) + b2   # (num_classes,)
else:
    # If single Dense directly after GAP -> no penultimate; W2 is already channels->classes
    W_combined, b_combined = W2, b2

# Save combined weights which are compatible with feature maps (channels x num_classes)
np.save(OUT_DIR / f"{CROP_NAME.lower()}_classifier_weights_combined_{PRE_TRAINED_MODEL}.npy", W_combined)
np.save(OUT_DIR / f"{CROP_NAME.lower()}_classifier_bias_combined_{PRE_TRAINED_MODEL}.npy", b_combined)

print("Saved combined classifier weights:", W_combined.shape, " bias:", b_combined.shape)


# 4) Convert wrapper to TFLite; try normal, fallback to SELECT_TF_OPS if needed
TFLITE_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
converter = tf.lite.TFLiteConverter.from_keras_model(wrapper_model)

# try conversion; if converter needs TF ops, convert with SELECT_TF_OPS
try:
    tflite_model = converter.convert()
    TFLITE_PATH.write_bytes(tflite_model)
    print(f"✅ TFLite (two-output) saved to: {TFLITE_PATH}")
except Exception as e:
    print("⚠️ First conversion failed — attempting with SELECT_TF_OPS. Error:", e)
    converter.target_spec.supported_ops = [
        tf.lite.OpsSet.TFLITE_BUILTINS,
        tf.lite.OpsSet.SELECT_TF_OPS
    ]
    tflite_model = converter.convert()
    TFLITE_PATH.write_bytes(tflite_model)
    print(f"✅ TFLite (two-output, SELECT_TF_OPS) saved to: {TFLITE_PATH}")

print("Done. Files in:", OUT_DIR.resolve())
```

```python
import tensorflow as tf
import numpy as np
import cv2
import os
from pathlib import Path
from tqdm import tqdm
import pandas as pd

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_recall_fscore_support,
    matthews_corrcoef,
)

preprocess_fn = tf.keras.applications.xception.preprocess_input

# ==============================
# CONFIG (edit these)
# ==============================
CROP_NAME = "Grape"
PRE_TRAINED_MODEL = "Xception_TrainTestVal"
OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")
SAVE_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/Evaluation")
# SAVE_DIR = Path("../Datasets/Evaluation")
OUT_DIR.mkdir(parents=True, exist_ok=True)
SAVE_DIR.mkdir(parents=True, exist_ok=True)
DATASET_PATH = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated") / CROP_NAME
# DATASET_PATH = Path("../Datasets/plant-village-dataset-updated-aggregated") / CROP_NAME
MODEL_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
LABELS_PATH = Path(f"{CROP_NAME.lower()}_disease_labels.txt")
IMG_SIZE = (224, 224)
SCALE_0_1 = False  # set True only if model expects [0,1] inputs

# ==============================
# LOAD LABELS
# ==============================
assert LABELS_PATH.exists(), f"Labels file not found: {LABELS_PATH}"
with open(LABELS_PATH, "r") as f:
    class_names = [line.strip() for line in f if line.strip()]

num_classes = len(class_names)
print(f"✅ Loaded {num_classes} classes from {LABELS_PATH.name}:")
print(class_names)

# ==============================
# LOAD TFLITE MODEL
# ==============================
assert MODEL_PATH.exists(), f"TFLite model not found: {MODEL_PATH}"
interpreter = tf.lite.Interpreter(model_path=str(MODEL_PATH))
interpreter.allocate_tensors()
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# ==============================
# HELPERS
# ==============================
def preprocess_image(img_path: Path):
    """Read image, convert to RGB, resize and optionally scale."""
    img = cv2.imread(str(img_path))
    if img is None:
        raise ValueError(f"Failed to read image: {img_path}")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, IMG_SIZE)
    img = img.astype(np.float32)
    if SCALE_0_1:
        img = img / 255.0
    else:
        # Use the Keras preprocess_input for the backbone (e.g. Xception -> [-1,1])
        img = preprocess_fn(img)
    img = np.expand_dims(img, axis=0)  # [1, H, W, 3]
    return img

def predict_idx(img_path: Path) -> int:
    """Predict class index for a single image using the TFLite interpreter."""
    x = preprocess_image(img_path)
    interpreter.set_tensor(input_details[0]['index'], x)
    interpreter.invoke()
    out = interpreter.get_tensor(output_details[-1]['index'])[0]  # robust to multi-output tflite
    out = np.asarray(out).astype(np.float32)
    return int(np.argmax(out))

# ==============================
# COLLECT FILES + PER-FOLDER COUNTS
# ==============================
all_files = []
per_folder_counts = {}
label_to_index = {name: i for i, name in enumerate(class_names)}

for disease_name in class_names:
    folder = DATASET_PATH / disease_name
    if not folder.exists():
        print(f"⚠️ Skipping missing folder (in dataset): {folder}")
        per_folder_counts[disease_name] = 0
        continue
    count = 0
    for root, _, files in os.walk(folder):
        for f in files:
            if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.webp', '.jfif', '.tif', '.tiff')):
                all_files.append((Path(root) / f, label_to_index[disease_name]))
                count += 1
    per_folder_counts[disease_name] = count
    print(f"📁 Folder '{disease_name}': {count} images")

print(f"\n🔍 Found {len(all_files)} images for evaluation in total.")

# ==============================
# EVALUATE WITH PROGRESS BAR
# ==============================
y_true, y_pred = [], []
for img_path, true_idx in tqdm(all_files, desc=f"Evaluating {CROP_NAME}", ncols=100):
    pred_idx = predict_idx(img_path)
    y_true.append(true_idx)
    y_pred.append(pred_idx)

y_true = np.array(y_true, dtype=np.int32)
y_pred = np.array(y_pred, dtype=np.int32)

# ==============================
# CONFUSION MATRIX (print first as requested)
# ==============================
print("\n🧮 Confusion Matrix (rows=True, cols=Pred):")
cm = confusion_matrix(y_true, y_pred, labels=list(range(num_classes)))
print(cm)

# ==============================
# classification_report dict & TP/FP/FN/TN
# ==============================
report_dict = classification_report(y_true, y_pred, target_names=class_names, output_dict=True, zero_division=0)

TP = cm.diagonal().astype(np.float64)
FP = cm.sum(axis=0) - TP
FN = cm.sum(axis=1) - TP
TN = cm.sum() - (TP + FP + FN)

eps = 1e-12  # avoid division by zero

# Build per-class rows by combining sklearn report + manual fields
rows = []
supports = []
specificities = []
gdrs = []

for i, name in enumerate(class_names):
    prec = float(report_dict[name]["precision"])
    rec = float(report_dict[name]["recall"])   # sensitivity
    f1 = float(report_dict[name]["f1-score"])
    support = int(report_dict[name]["support"])
    supports.append(support)

    spec = float(TN[i] / (TN[i] + FP[i] + eps))
    gdr_val = float(np.sqrt(rec * spec))
    gdrs.append(gdr_val)
    specificities.append(spec)

    per_class_acc = float((TP[i] + TN[i]) / (TP[i] + TN[i] + FP[i] + FN[i] + eps))

    rows.append({
        "class": name,
        "support": support,
        "TP": int(TP[i]),
        "FP": int(FP[i]),
        "FN": int(FN[i]),
        "TN": int(TN[i]),
        "precision": prec,
        "recall (sensitivity)": rec,
        "specificity": spec,
        "f1": f1,
        "gdr (g-mean)": gdr_val,
        "per-class accuracy": per_class_acc,
    })

metrics_df = pd.DataFrame(rows)

# Round float columns to 4 decimal places for compact display and CSV
float_cols = ["precision", "recall (sensitivity)", "specificity", "f1", "gdr (g-mean)", "per-class accuracy"]
metrics_df[float_cols] = metrics_df[float_cols].round(4)

# Reorder columns for compactness
metrics_df = metrics_df[
    ["class", "support", "TP", "FP", "FN", "TN",
     "precision", "recall (sensitivity)", "specificity",
     "f1", "gdr (g-mean)", "per-class accuracy"]
]

# ==============================
# PRINT per-class table compactly (rounded)
# ==============================
print("\n📋 Per-class metrics (combined from classification_report + extras):")
# Use to_string with float_format to ensure 4 decimal places in console
print(metrics_df.to_string(index=False, float_format="{:.4f}".format))

# ==============================
# Macro & Weighted averages
# - precision/recall/f1 from classification_report dict (if present)
# - specificity/gdr: compute macro (mean) and weighted (support-weighted)
# ==============================
macro = report_dict.get("macro avg", {})
weighted = report_dict.get("weighted avg", {})

supports_arr = np.array(supports, dtype=np.float64)
macro_specificity = float(np.mean(specificities))
macro_gdr = float(np.mean(gdrs))
weighted_specificity = float(np.average(specificities, weights=supports_arr))
weighted_gdr = float(np.average(gdrs, weights=supports_arr))

# Safely get macro/weighted precision/recall/f1; fallback to sklearn calc if missing
if macro:
    macro_prec = macro.get("precision", np.nan)
    macro_rec = macro.get("recall", np.nan)
    macro_f1 = macro.get("f1-score", np.nan)
else:
    macro_prec, macro_rec, macro_f1, _ = precision_recall_fscore_support(y_true, y_pred, average='macro', zero_division=0)

if weighted:
    weighted_prec = weighted.get("precision", np.nan)
    weighted_rec = weighted.get("recall", np.nan)
    weighted_f1 = weighted.get("f1-score", np.nan)
else:
    weighted_prec, weighted_rec, weighted_f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=0)

# Round summary values to 4 decimals
def r(x): return float(np.round(x, 4)) if not (x is None or (isinstance(x, float) and np.isnan(x))) else np.nan

macro_prec, macro_rec, macro_f1 = r(macro_prec), r(macro_rec), r(macro_f1)
weighted_prec, weighted_rec, weighted_f1 = r(weighted_prec), r(weighted_rec), r(weighted_f1)
macro_specificity, weighted_specificity = r(macro_specificity), r(weighted_specificity)
macro_gdr, weighted_gdr = r(macro_gdr), r(weighted_gdr)

# ==============================
# Remaining overall metrics and save CSV
# ==============================
acc = accuracy_score(y_true, y_pred)
mcc = matthews_corrcoef(y_true, y_pred)

print("\n🔢 Overall metrics:")
print(f"Overall accuracy: {acc:.4f}")
print(f"MCC (Matthews Correlation Coefficient): {mcc:.4f}")

# Save rounded per-class CSV for thesis tables/plots
csv_out = SAVE_DIR / f"{CROP_NAME.lower()}_evaluation_metrics_{PRE_TRAINED_MODEL}.csv"
metrics_df.to_csv(csv_out, index=False, float_format="%.4f")
np.save(SAVE_DIR / f"{CROP_NAME.lower()}_confusion_matrix_{PRE_TRAINED_MODEL}.npy", cm)

print(f"\nSaved per-class metrics CSV to: {csv_out}")
print(f"Saved confusion matrix (.npy) to: {SAVE_DIR / f'{CROP_NAME.lower()}_confusion_matrix_{PRE_TRAINED_MODEL}.npy'}")

# Final thesis-ready summary (explicit precision and recall included)
print("\n=== Thesis-ready summary ===")
print(f"Overall accuracy: {acc:.4f}")
print(f"Macro  - Precision: {macro_prec:.4f} | Recall: {macro_rec:.4f} | F1: {macro_f1:.4f} | Specificity: {macro_specificity:.4f} | GDR (g-mean): {macro_gdr:.4f}")
print(f"Weighted - Precision: {weighted_prec:.4f} | Recall: {weighted_rec:.4f} | F1: {weighted_f1:.4f} | Specificity: {weighted_specificity:.4f} | GDR (g-mean): {weighted_gdr:.4f}")
print(f"MCC: {mcc:.4f}")
print("Per-class metrics saved to CSV and printed above.")
```

```python
# Single notebook cell: Generate Grad-CAM images (5 per class) and display in grid (5 per row)
# -----------------------------------------------------------------------------
import tensorflow as tf
import numpy as np
import cv2
import os
import random
from pathlib import Path
from tqdm import tqdm
import matplotlib.pyplot as plt
import math

# ---------------------- USER CONFIG ----------------------
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
preprocess_fn = tf.keras.applications.xception.preprocess_input

CROP_NAME = "Grape"   # e.g. "Apple", "Tomato"
PRE_TRAINED_MODEL = "Xception_TrainTestVal"
OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")   # where tflite + weights .npy live and where outputs will be written
# Local
# DATASET_ROOT = Path("../Datasets/plant-village-dataset-updated-aggregated")
# Colab
DATASET_ROOT = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated")
IMG_SIZE = (224, 224)   # model input size
IMAGES_PER_CLASS = 5    # how many random images per class to create heatmaps for
SCALE_0_1 = False       # set True if model expects inputs scaled to [0,1]; False to use 0..255 floats
VERBOSE = True
# ---------------------------------------------------------

# Derived paths
DATASET_PATH = DATASET_ROOT / CROP_NAME
TFLITE_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
WEIGHTS_PATH = OUT_DIR / f"{CROP_NAME.lower()}_classifier_weights_combined_{PRE_TRAINED_MODEL}.npy"
BIAS_PATH = OUT_DIR / f"{CROP_NAME.lower()}_classifier_bias_combined_{PRE_TRAINED_MODEL}.npy"
LABELS_PATH = Path(f"{CROP_NAME.lower()}_disease_labels.txt")
SAVE_DIR = OUT_DIR / "gradcam" / CROP_NAME
SAVE_DIR.mkdir(parents=True, exist_ok=True)

# Sanity checks
assert DATASET_PATH.exists(), f"Dataset folder not found: {DATASET_PATH}"
assert TFLITE_PATH.exists(), f"TFLite file not found: {TFLITE_PATH}"
assert WEIGHTS_PATH.exists(), f"Weights file not found: {WEIGHTS_PATH}"
assert LABELS_PATH.exists(), f"Labels file not found: {LABELS_PATH}"

# Load labels
with open(LABELS_PATH, "r") as f:
    class_names = [ln.strip() for ln in f if ln.strip()]
num_classes = len(class_names)
if VERBOSE:
    print(f"Loaded {num_classes} labels: {class_names}")

# Load classifier weights & bias
W = np.load(WEIGHTS_PATH)   # expected shape (channels, num_classes) if GAP -> Dense
b = None
if BIAS_PATH.exists():
    b = np.load(BIAS_PATH)

if VERBOSE:
    print("Classifier weight shape:", W.shape, " bias shape:", None if b is None else b.shape)

# Load TFLite interpreter
interpreter = tf.lite.Interpreter(model_path=str(TFLITE_PATH))
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
if VERBOSE:
    print("TFLite Input details:", input_details)
    print("TFLite Output details:", output_details)

# Identify which output index is feature maps and which is probabilities:
feat_idx = None
prob_idx = None
for od in output_details:
    shape = od["shape"]
    if len(shape) == 4:
        feat_idx = od["index"]
    elif len(shape) == 2:
        prob_idx = od["index"]

if feat_idx is None or prob_idx is None:
    raise RuntimeError("Couldn't identify feature-map and prob outputs from TFLite. Check model outputs.")

if VERBOSE:
    print("Using feature-output index:", feat_idx, "prob-output index:", prob_idx)

# Helper: image preproc (assumes model contains preprocess step, so pass raw 0..255 floats)
def preprocess_image(path: Path):
    img_bgr = cv2.imread(str(path))
    if img_bgr is None:
        raise ValueError(f"Failed to load image {path}")
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    resized = cv2.resize(img_rgb, IMG_SIZE)
    arr = resized.astype(np.float32)
    if SCALE_0_1:
        arr = arr / 255.0
    else:
        arr = preprocess_fn(arr)
    return arr, resized  # arr used as model input, resized is used for overlay visualization (RGB uint8)

# Helper: compute grad-cam heatmap from feature maps and classifier weights
def compute_gradcam_from_featuremaps(feature_maps: np.ndarray, class_id: int, weights: np.ndarray):
    """
    feature_maps: (1, H, W, C) float32
    weights: (C, num_classes) or (num_classes, C)
    returns heatmap HxW normalized [0,1]
    """
    fm = feature_maps[0]   # H x W x C
    # find correct weight vector for class
    if weights.shape[0] == fm.shape[2] and weights.shape[1] == num_classes:
        w_k = weights[:, class_id]    # (C,)
    elif weights.shape[1] == fm.shape[2] and weights.shape[0] == num_classes:
        w_k = weights[class_id, :]
    else:
        w_try = weights.T
        if w_try.shape[0] == fm.shape[2]:
            w_k = w_try[:, class_id]
        else:
            raise ValueError(f"Classifier weights shape {weights.shape} incompatible with feature channels {fm.shape[2]}")

    heatmap = np.tensordot(fm, w_k, axes=([2], [0]))   # shape (H, W)
    heatmap = np.maximum(heatmap, 0)
    if np.max(heatmap) > 0:
        heatmap = heatmap / np.max(heatmap)
    else:
        heatmap = np.zeros_like(heatmap)
    return heatmap.astype(np.float32)

# Helper: overlay heatmap on image
def overlay_heatmap_on_image(orig_rgb, heatmap, alpha=0.45, colormap=cv2.COLORMAP_JET):
    hm_u8 = np.uint8(255 * heatmap)
    hm_color = cv2.applyColorMap(hm_u8, colormap)         # BGR
    hm_color = cv2.cvtColor(hm_color, cv2.COLOR_BGR2RGB)  # to RGB
    hm_resized = cv2.resize(hm_color, (orig_rgb.shape[1], orig_rgb.shape[0]))
    overlay = (0.5 * orig_rgb + alpha * hm_resized).astype(np.uint8)
    return overlay

# Collect up to IMAGES_PER_CLASS random images per class
samples = []
for cls in class_names:
    folder = DATASET_PATH / cls
    if not folder.exists():
        print(f"⚠️ Missing folder {folder}, skipping")
        continue
    files = [Path(folder) / f for f in os.listdir(folder) if f.lower().endswith(('.jpg','.jpeg','.png','.bmp','.tif','.tiff'))]
    if len(files) == 0:
        continue
    chosen = random.sample(files, k=min(IMAGES_PER_CLASS, len(files)))
    for p in chosen:
        samples.append((cls, p))

if VERBOSE:
    print(f"Selected {len(samples)} total images ({IMAGES_PER_CLASS} per class where available).")

# Run tflite inference, compute heatmaps, save overlays
records = []
for cls, img_path in tqdm(samples, desc="Generating Grad-CAM", ncols=100):
    img_arr, resized_rgb = preprocess_image(img_path)
    input_tensor = np.expand_dims(img_arr, axis=0).astype(np.float32)

    interpreter.set_tensor(input_details[0]['index'], input_tensor)
    interpreter.invoke()
    feat = interpreter.get_tensor(feat_idx)   # (1,Hf,Wf,C)
    probs = interpreter.get_tensor(prob_idx)  # (1,num_classes)
    probs = np.asarray(probs).astype(np.float32)[0]
    pred_idx = int(np.argmax(probs))
    pred_label = class_names[pred_idx] if pred_idx < len(class_names) else str(pred_idx)
    confidence = float(probs[pred_idx])

    heatmap = compute_gradcam_from_featuremaps(feat, pred_idx, W)
    overlay = overlay_heatmap_on_image(resized_rgb, heatmap, alpha=0.5)

    class_folder = SAVE_DIR / cls
    class_folder.mkdir(parents=True, exist_ok=True)

    base_name = img_path.stem
    out_heatmap_path = class_folder / f"{base_name}_heatmap.png"
    out_overlay_path = class_folder / f"{base_name}_overlay.png"
    out_orig_path = class_folder / f"{base_name}_orig.png"

    cv2.imwrite(str(out_heatmap_path), (255 * cv2.resize(heatmap, (IMG_SIZE[0], IMG_SIZE[1]))).astype(np.uint8))
    cv2.imwrite(str(out_overlay_path), cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))
    cv2.imwrite(str(out_orig_path), cv2.cvtColor(resized_rgb, cv2.COLOR_RGB2BGR))

    records.append({
        "true_class": cls,
        "image_path": str(img_path),
        "predicted_label": pred_label,
        "confidence": confidence,
        "overlay": str(out_overlay_path),
        "orig": str(out_orig_path),
        "heatmap": str(out_heatmap_path)
    })

print(f"\nSaved {len(records)} Grad-CAM images to: {SAVE_DIR}")

# ------------------ DISPLAY: grid with 5 images per row ------------------
# ------------------ DISPLAY: grid with 5 images per row, no horizontal scroll, dark bold titles ------------------
if len(records) == 0:
    print("No records to display.")
else:
    per_row = 5                       # show exactly 5 images per row
    n = len(records)
    rows = math.ceil(n / per_row)

    # Prevent horizontal scrolling by capping figure width (in inches).
    # Increase/decrease max_width_inch if you know your display width.
    max_width_inch = 15
    cell_size_inch = 2.8              # size per image cell (adjust if you want larger/smaller)
    fig_w = min(per_row * cell_size_inch, max_width_inch)
    fig_h = rows * cell_size_inch

    fig, axes = plt.subplots(rows, per_row, figsize=(fig_w, fig_h))
    # ensure axes is 2D array even if rows==1
    if rows == 1:
        axes = np.expand_dims(axes, axis=0)
    axes = np.array(axes).reshape(rows, per_row)

    for idx, rec in enumerate(records):
        r = idx // per_row
        c = idx % per_row
        ax = axes[r, c]
        img_bgr = cv2.imread(rec["overlay"])
        if img_bgr is None:
            img_bgr = cv2.imread(rec["orig"])  # fallback
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
        ax.imshow(img_rgb)
        title = f"T:{rec['true_class']}  P:{rec['predicted_label']} ({rec['confidence']*100:.1f}%)"
        ax.set_title(title, fontsize=9, color='black', fontweight='bold')   # dark bold font
        ax.axis("off")

    # hide unused subplots
    for idx in range(n, rows * per_row):
        r = idx // per_row
        c = idx % per_row
        axes[r, c].axis("off")

    plt.tight_layout()
    plt.show()

```

```python
import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.metrics import classification_report, confusion_matrix
import os
import shutil
from collections import defaultdict
import math
from collections import Counter
import pandas as pd

# Random seed setup
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

# Dataset path local/colab
# Colab
DATASET_PATH = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated")
# Local
# DATASET_PATH = Path("../Datasets/plant-village-dataset-updated-aggregated")

# Change crop name here
CROP_NAME = "Grape"
CROP_PATH = DATASET_PATH / CROP_NAME
IMG_SIZE = (224, 224)
# CPU
# BATCH_SIZE = 32
# GPU
BATCH_SIZE = 64
EPOCHS = 30
PRE_TRAINED_MODEL = "DenseNet121_TrainTestVal"
DROPOUT = 0.3
LR_RATE = 1e-4

OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# add this (minimal) — DenseNet121 preprocessing
preprocess_fn = tf.keras.applications.densenet.preprocess_input


# Data Loader (70% / 15% / 15%)
# Count per class in each split
def count_per_class(file_paths, class_names):
    labels = [Path(f).parent.name for f in file_paths]
    counts = Counter(labels)
    return {cls: counts.get(cls, 0) for cls in class_names}

# Gather file list to compute exact counts
all_files = []
for root, _, files in os.walk(CROP_PATH):
    for f in files:
        if f.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp")):
            all_files.append(Path(root) / f)
N = len(all_files)
if N == 0:
    raise RuntimeError(f"No images found under {CROP_PATH}")

train_count = int(N * 0.70)
val_count = int(N * 0.15)
# Validate sum
test_count = N - train_count - val_count

print(f"Total images: {N} -> train: {train_count}, val: {val_count}, test: {test_count}")

# Use image_dataset_from_directory with validation_split=0.3:
rest_split = 0.3

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    CROP_PATH,
    validation_split=rest_split,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True
)

rest_ds = tf.keras.preprocessing.image_dataset_from_directory(
    CROP_PATH,
    validation_split=rest_split,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,   # batched
    shuffle=True
)

class_names = train_ds.class_names
num_classes = len(class_names)

# Print the count of images per disease for Train, Test, and Val
train_files = train_ds.file_paths
rest_files = [p for p in rest_ds.file_paths]
val_files = rest_files[:val_count]
test_files = rest_files[val_count:val_count+test_count]

train_counts = count_per_class(train_files, class_names)
val_counts = count_per_class(val_files, class_names)
test_counts = count_per_class(test_files, class_names)

# Create DataFrame for thesis table
df = pd.DataFrame({
    "Train": train_counts,
    "Val": val_counts,
    "Test": test_counts
})
print("\n Per-class image distribution:\n")
print(df)
column_sums = df.sum()

print("\n--------------------")
print("Sum of each column:")
print(column_sums)

# Flatten rest_ds and split deterministically by counts
rest_ds = rest_ds.unbatch()
val_ds = rest_ds.take(val_count)
test_ds = rest_ds.skip(val_count).take(test_count)

# Rebatch val & test for training pipeline; we'll also create finite eval datasets below
val_ds = val_ds.batch(BATCH_SIZE)
test_ds = test_ds.batch(BATCH_SIZE)

# Derive class_names from train_ds
print("\n--------------------")
print(f"✅ Training {CROP_NAME} model with {num_classes} classes")
print("Classes:", class_names)
train_ds_len = tf.data.experimental.cardinality(train_ds)
print(f"Total images: {N} -> train data per batch: {train_ds_len.numpy()}")

# Save class labels
with open(f"{CROP_NAME.lower()}_disease_labels.txt", "w") as f:
    for name in class_names:
        f.write(name + "\n")

# Data augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
    tf.keras.layers.RandomContrast(0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
    tf.keras.layers.GaussianNoise(0.05),
    tf.keras.layers.RandomBrightness(factor=0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
])

AUTOTUNE = tf.data.AUTOTUNE


# Pipelines

# TRAINING pipeline: cache -> shuffle -> augment -> preprocess -> repeat -> prefetch
train_ds = (
    train_ds
    .cache()
    .shuffle(1000)
    .map(lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=AUTOTUNE)
    .map(lambda x, y: (preprocess_fn(tf.cast(x, tf.float32)), y), num_parallel_calls=AUTOTUNE)   # <-- added preprocessing
    .repeat()   # <-- keep repeat so fit() can run many epochs
    .prefetch(buffer_size=AUTOTUNE)
)

# VALIDATION for training (use a repeated version for validation during fit)
val_ds_for_fit = (
    val_ds
    .cache()
    .map(lambda x, y: (preprocess_fn(tf.cast(x, tf.float32)), y), num_parallel_calls=AUTOTUNE)
    .repeat()
    .prefetch(buffer_size=AUTOTUNE)
)

val_ds_for_eval = (
    val_ds
    .cache()
    .map(lambda x, y: (preprocess_fn(tf.cast(x, tf.float32)), y), num_parallel_calls=AUTOTUNE)
    .prefetch(buffer_size=AUTOTUNE)
)

# VAL pipeline used for final evaluation remains finite (val_ds)
# TEST pipeline remains finite
test_ds = (
    test_ds
    .cache()
    .map(lambda x, y: (preprocess_fn(tf.cast(x, tf.float32)), y), num_parallel_calls=AUTOTUNE)
    .prefetch(buffer_size=AUTOTUNE)
)


# compute steps for training loops
steps_per_epoch = max(1, train_count // BATCH_SIZE)
validation_steps = max(1, val_count // BATCH_SIZE)
print(f"steps_per_epoch={steps_per_epoch}, validation_steps={validation_steps}")

# Model Building
base_model = tf.keras.applications.DenseNet121(
    input_shape=IMG_SIZE + (3,),
    include_top=False,
    weights="imagenet",
    # include_preprocessing=True
)
base_model.trainable = False  # freeze backbone initially

# Input -> backbone -> head (we reuse base_model.input/output)
# x = base_model.output
# x = tf.keras.layers.GlobalAveragePooling2D()(x)
# x = tf.keras.layers.Dropout(0.3)(x)
# outputs = tf.keras.layers.Dense(num_classes, activation="softmax", name="predictions")(x)

x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(256, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(LR_RATE))(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dropout(DROPOUT)(x)
outputs = tf.keras.layers.Dense(num_classes, activation="softmax", name="predictions")(x)

model = tf.keras.Model(base_model.input, outputs, name=f"{PRE_TRAINED_MODEL}_{CROP_NAME}")

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# Training
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss", patience=7, restore_best_weights=True
)

checkpoint = tf.keras.callbacks.ModelCheckpoint(
    f"best_{CROP_NAME.lower()}_disease_model_{PRE_TRAINED_MODEL}.keras",
    save_best_only=True,
    monitor="val_accuracy",
    mode="max"
)

history = model.fit(
    train_ds,
    validation_data=val_ds_for_fit,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# Fine-tuning: unfreeze some top layers of base_model
base_model.trainable = True
fine_tune_at = int(len(base_model.layers) * 0.75)
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_fine = model.fit(
    train_ds,
    validation_data=val_ds_for_fit,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# Evaluate on validation
results = model.evaluate(val_ds_for_eval)
print("Metrics - ")
for name, value in zip(model.metrics_names, results):
    print(f"{name}: {value:.4f}")

# Plots: Accuracy & Loss
plt.figure(figsize=(10, 6))
plt.plot(history.history.get('accuracy', []) + history_fine.history.get('accuracy', []), label="Training Accuracy")
plt.plot(history.history.get('val_accuracy', []) + history_fine.history.get('val_accuracy', []), label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title(f"{CROP_NAME} - Training & Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(history.history.get('loss', []) + history_fine.history.get('loss', []), label="Training Loss")
plt.plot(history.history.get('val_loss', []) + history_fine.history.get('val_loss', []), label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title(f"{CROP_NAME} - Training & Validation Loss")
plt.legend()
plt.grid(True)
plt.show()

# Validation Evaluation: Classification Report & Confusion Matrix on VAL set
# ==============================
y_true = []
y_pred = []

for batch_images, batch_labels in val_ds_for_eval:
    preds = model.predict(batch_images, verbose=0)
    y_true.extend(batch_labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

# Safety: ensure lengths align
if len(class_names) != (int(max(max(y_true, default=0), max(y_pred, default=0))) + 1):
    print("Warning: class_names length does not match labels. Recomputing class_names from folders.")
    class_names = sorted([d.name for d in CROP_PATH.iterdir() if d.is_dir()])
    num_classes = len(class_names)
    print("New class_names:", class_names)

y_true = np.array(y_true, dtype=np.int64)
y_pred = np.array(y_pred, dtype=np.int64)

acc = float(np.mean(y_true == y_pred))
print(f"\n Overall accuracy: {acc*100:.2f}%")

print("Validation Classification Report:")
print(classification_report(y_true, y_pred,labels=list(range(num_classes)), target_names=class_names, zero_division=0))

cm = confusion_matrix(y_true, y_pred, labels=range(len(class_names)))
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"{CROP_NAME} - Validation Confusion Matrix")
plt.show()

# Testing Evaluation: classification report & confusion matrix on TEST set
y_true_test = []
y_pred_test = []

for batch_images, batch_labels in test_ds:
    preds = model.predict(batch_images, verbose=0)
    y_true_test.extend(batch_labels.numpy())
    y_pred_test.extend(np.argmax(preds, axis=1))

y_true_test = np.array(y_true_test, dtype=np.int64)
y_pred_test = np.array(y_pred_test, dtype=np.int64)

acc = float(np.mean(y_true_test == y_pred_test))
print(f"\nOverall accuracy: {acc*100:.2f}%")

print("Test Classification Report:")
print(classification_report(y_true_test, y_pred_test, target_names=class_names, zero_division=0))

cm_test = confusion_matrix(y_true_test, y_pred_test, labels=range(len(class_names)))
plt.figure(figsize=(8, 6))
sns.heatmap(cm_test, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"{CROP_NAME} - Test Confusion Matrix")
plt.show()

# EXPORT TFLITE wrapper for Grad-CAM (two outputs)
# 1) Find last convolutional layer
def find_last_conv_layer(model: tf.keras.Model):
    for layer in reversed(model.layers):
        lname = layer.__class__.__name__.lower()
        if ("conv" in lname) or ("depthwiseconv" in lname) or ("separableconv" in lname):
            return layer.name
    raise ValueError("No convolutional layer found in the model. Can't create CAM without conv layers.")

last_conv_name = find_last_conv_layer(model)
print("Last conv layer:", last_conv_name)

# 2) Build wrapper model (preds, feature_maps)
last_conv_layer = model.get_layer(last_conv_name)
wrapper_model = tf.keras.Model(inputs=model.input, outputs=[model.output, last_conv_layer.output],
                               name=f"{model.name}_with_featuremaps")
wrapper_model.summary(line_length=120)

# 3) Save classifier Dense weights (for CAM on-device)
# Find dense layers in the head (expect two: penultimate and classifier)
dense_layers = [l for l in model.layers if isinstance(l, tf.keras.layers.Dense)]
if len(dense_layers) < 1:
    raise ValueError("No Dense layer found in the model head.")

# if there are >=2, last is classifier, second-last is penultimate
classifier_layer = dense_layers[-1]
W2, b2 = classifier_layer.get_weights()   # shape (units_prev, num_classes), (num_classes,)
if len(dense_layers) >= 2:
    penult_layer = dense_layers[-2]
    W1, b1 = penult_layer.get_weights()   # shape (channels, penult_units), (penult_units,)
    # Combined mapping from GAP channels -> classes
    # W_combined: (channels, num_classes)
    W_combined = np.matmul(W1, W2)        # (channels, num_classes)
    # Combined bias
    b_combined = np.matmul(b1, W2) + b2   # (num_classes,)
else:
    # If single Dense directly after GAP -> no penultimate; W2 is already channels->classes
    W_combined, b_combined = W2, b2

# Save combined weights which are compatible with feature maps (channels x num_classes)
np.save(OUT_DIR / f"{CROP_NAME.lower()}_classifier_weights_combined_{PRE_TRAINED_MODEL}.npy", W_combined)
np.save(OUT_DIR / f"{CROP_NAME.lower()}_classifier_bias_combined_{PRE_TRAINED_MODEL}.npy", b_combined)

print("Saved combined classifier weights:", W_combined.shape, " bias:", b_combined.shape)


# 4) Convert wrapper to TFLite; try normal, fallback to SELECT_TF_OPS if needed
TFLITE_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
converter = tf.lite.TFLiteConverter.from_keras_model(wrapper_model)

# try conversion; if converter needs TF ops, convert with SELECT_TF_OPS
try:
    tflite_model = converter.convert()
    TFLITE_PATH.write_bytes(tflite_model)
    print(f"✅ TFLite (two-output) saved to: {TFLITE_PATH}")
except Exception as e:
    print("⚠️ First conversion failed — attempting with SELECT_TF_OPS. Error:", e)
    converter.target_spec.supported_ops = [
        tf.lite.OpsSet.TFLITE_BUILTINS,
        tf.lite.OpsSet.SELECT_TF_OPS
    ]
    tflite_model = converter.convert()
    TFLITE_PATH.write_bytes(tflite_model)
    print(f"✅ TFLite (two-output, SELECT_TF_OPS) saved to: {TFLITE_PATH}")

print("Done. Files in:", OUT_DIR.resolve())
```

```python
import tensorflow as tf
import numpy as np
import cv2
import os
from pathlib import Path
from tqdm import tqdm
import pandas as pd

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_recall_fscore_support,
    matthews_corrcoef,
)

preprocess_fn = tf.keras.applications.densenet.preprocess_input

# ==============================
# CONFIG (edit these)
# ==============================
CROP_NAME = "Grape"
PRE_TRAINED_MODEL = "DenseNet121_TrainTestVal"
OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")
SAVE_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/Evaluation")
# SAVE_DIR = Path("../Datasets/Evaluation")
OUT_DIR.mkdir(parents=True, exist_ok=True)
SAVE_DIR.mkdir(parents=True, exist_ok=True)
DATASET_PATH = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated") / CROP_NAME
# DATASET_PATH = Path("../Datasets/plant-village-dataset-updated-aggregated") / CROP_NAME
MODEL_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
LABELS_PATH = Path(f"{CROP_NAME.lower()}_disease_labels.txt")
IMG_SIZE = (224, 224)
SCALE_0_1 = False  # set True only if model expects [0,1] inputs

# ==============================
# LOAD LABELS
# ==============================
assert LABELS_PATH.exists(), f"Labels file not found: {LABELS_PATH}"
with open(LABELS_PATH, "r") as f:
    class_names = [line.strip() for line in f if line.strip()]

num_classes = len(class_names)
print(f"✅ Loaded {num_classes} classes from {LABELS_PATH.name}:")
print(class_names)

# ==============================
# LOAD TFLITE MODEL
# ==============================
assert MODEL_PATH.exists(), f"TFLite model not found: {MODEL_PATH}"
interpreter = tf.lite.Interpreter(model_path=str(MODEL_PATH))
interpreter.allocate_tensors()
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# ==============================
# HELPERS
# ==============================
def preprocess_image(img_path: Path):
    """Read image, convert to RGB, resize and optionally scale."""
    img = cv2.imread(str(img_path))
    if img is None:
        raise ValueError(f"Failed to read image: {img_path}")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, IMG_SIZE)
    img = img.astype(np.float32)
    if SCALE_0_1:
        img = img / 255.0
    else:
        # Use the Keras preprocess_input for the backbone (e.g. Xception -> [-1,1])
        img = preprocess_fn(img)
    img = np.expand_dims(img, axis=0)  # [1, H, W, 3]
    return img

def predict_idx(img_path: Path) -> int:
    """Predict class index for a single image using the TFLite interpreter."""
    x = preprocess_image(img_path)
    interpreter.set_tensor(input_details[0]['index'], x)
    interpreter.invoke()
    out = interpreter.get_tensor(output_details[-1]['index'])[0]  # robust to multi-output tflite
    out = np.asarray(out).astype(np.float32)
    return int(np.argmax(out))

# ==============================
# COLLECT FILES + PER-FOLDER COUNTS
# ==============================
all_files = []
per_folder_counts = {}
label_to_index = {name: i for i, name in enumerate(class_names)}

for disease_name in class_names:
    folder = DATASET_PATH / disease_name
    if not folder.exists():
        print(f"⚠️ Skipping missing folder (in dataset): {folder}")
        per_folder_counts[disease_name] = 0
        continue
    count = 0
    for root, _, files in os.walk(folder):
        for f in files:
            if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.webp', '.jfif', '.tif', '.tiff')):
                all_files.append((Path(root) / f, label_to_index[disease_name]))
                count += 1
    per_folder_counts[disease_name] = count
    print(f"📁 Folder '{disease_name}': {count} images")

print(f"\n🔍 Found {len(all_files)} images for evaluation in total.")

# ==============================
# EVALUATE WITH PROGRESS BAR
# ==============================
y_true, y_pred = [], []
for img_path, true_idx in tqdm(all_files, desc=f"Evaluating {CROP_NAME}", ncols=100):
    pred_idx = predict_idx(img_path)
    y_true.append(true_idx)
    y_pred.append(pred_idx)

y_true = np.array(y_true, dtype=np.int32)
y_pred = np.array(y_pred, dtype=np.int32)

# ==============================
# CONFUSION MATRIX (print first as requested)
# ==============================
print("\n🧮 Confusion Matrix (rows=True, cols=Pred):")
cm = confusion_matrix(y_true, y_pred, labels=list(range(num_classes)))
print(cm)

# ==============================
# classification_report dict & TP/FP/FN/TN
# ==============================
report_dict = classification_report(y_true, y_pred, target_names=class_names, output_dict=True, zero_division=0)

TP = cm.diagonal().astype(np.float64)
FP = cm.sum(axis=0) - TP
FN = cm.sum(axis=1) - TP
TN = cm.sum() - (TP + FP + FN)

eps = 1e-12  # avoid division by zero

# Build per-class rows by combining sklearn report + manual fields
rows = []
supports = []
specificities = []
gdrs = []

for i, name in enumerate(class_names):
    prec = float(report_dict[name]["precision"])
    rec = float(report_dict[name]["recall"])   # sensitivity
    f1 = float(report_dict[name]["f1-score"])
    support = int(report_dict[name]["support"])
    supports.append(support)

    spec = float(TN[i] / (TN[i] + FP[i] + eps))
    gdr_val = float(np.sqrt(rec * spec))
    gdrs.append(gdr_val)
    specificities.append(spec)

    per_class_acc = float((TP[i] + TN[i]) / (TP[i] + TN[i] + FP[i] + FN[i] + eps))

    rows.append({
        "class": name,
        "support": support,
        "TP": int(TP[i]),
        "FP": int(FP[i]),
        "FN": int(FN[i]),
        "TN": int(TN[i]),
        "precision": prec,
        "recall (sensitivity)": rec,
        "specificity": spec,
        "f1": f1,
        "gdr (g-mean)": gdr_val,
        "per-class accuracy": per_class_acc,
    })

metrics_df = pd.DataFrame(rows)

# Round float columns to 4 decimal places for compact display and CSV
float_cols = ["precision", "recall (sensitivity)", "specificity", "f1", "gdr (g-mean)", "per-class accuracy"]
metrics_df[float_cols] = metrics_df[float_cols].round(4)

# Reorder columns for compactness
metrics_df = metrics_df[
    ["class", "support", "TP", "FP", "FN", "TN",
     "precision", "recall (sensitivity)", "specificity",
     "f1", "gdr (g-mean)", "per-class accuracy"]
]

# ==============================
# PRINT per-class table compactly (rounded)
# ==============================
print("\n📋 Per-class metrics (combined from classification_report + extras):")
# Use to_string with float_format to ensure 4 decimal places in console
print(metrics_df.to_string(index=False, float_format="{:.4f}".format))

# ==============================
# Macro & Weighted averages
# - precision/recall/f1 from classification_report dict (if present)
# - specificity/gdr: compute macro (mean) and weighted (support-weighted)
# ==============================
macro = report_dict.get("macro avg", {})
weighted = report_dict.get("weighted avg", {})

supports_arr = np.array(supports, dtype=np.float64)
macro_specificity = float(np.mean(specificities))
macro_gdr = float(np.mean(gdrs))
weighted_specificity = float(np.average(specificities, weights=supports_arr))
weighted_gdr = float(np.average(gdrs, weights=supports_arr))

# Safely get macro/weighted precision/recall/f1; fallback to sklearn calc if missing
if macro:
    macro_prec = macro.get("precision", np.nan)
    macro_rec = macro.get("recall", np.nan)
    macro_f1 = macro.get("f1-score", np.nan)
else:
    macro_prec, macro_rec, macro_f1, _ = precision_recall_fscore_support(y_true, y_pred, average='macro', zero_division=0)

if weighted:
    weighted_prec = weighted.get("precision", np.nan)
    weighted_rec = weighted.get("recall", np.nan)
    weighted_f1 = weighted.get("f1-score", np.nan)
else:
    weighted_prec, weighted_rec, weighted_f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=0)

# Round summary values to 4 decimals
def r(x): return float(np.round(x, 4)) if not (x is None or (isinstance(x, float) and np.isnan(x))) else np.nan

macro_prec, macro_rec, macro_f1 = r(macro_prec), r(macro_rec), r(macro_f1)
weighted_prec, weighted_rec, weighted_f1 = r(weighted_prec), r(weighted_rec), r(weighted_f1)
macro_specificity, weighted_specificity = r(macro_specificity), r(weighted_specificity)
macro_gdr, weighted_gdr = r(macro_gdr), r(weighted_gdr)

# ==============================
# Remaining overall metrics and save CSV
# ==============================
acc = accuracy_score(y_true, y_pred)
mcc = matthews_corrcoef(y_true, y_pred)

print("\n🔢 Overall metrics:")
print(f"Overall accuracy: {acc:.4f}")
print(f"MCC (Matthews Correlation Coefficient): {mcc:.4f}")

# Save rounded per-class CSV for thesis tables/plots
csv_out = SAVE_DIR / f"{CROP_NAME.lower()}_evaluation_metrics_{PRE_TRAINED_MODEL}.csv"
metrics_df.to_csv(csv_out, index=False, float_format="%.4f")
np.save(SAVE_DIR / f"{CROP_NAME.lower()}_confusion_matrix_{PRE_TRAINED_MODEL}.npy", cm)

print(f"\nSaved per-class metrics CSV to: {csv_out}")
print(f"Saved confusion matrix (.npy) to: {SAVE_DIR / f'{CROP_NAME.lower()}_confusion_matrix_{PRE_TRAINED_MODEL}.npy'}")

# Final thesis-ready summary (explicit precision and recall included)
print("\n=== Thesis-ready summary ===")
print(f"Overall accuracy: {acc:.4f}")
print(f"Macro  - Precision: {macro_prec:.4f} | Recall: {macro_rec:.4f} | F1: {macro_f1:.4f} | Specificity: {macro_specificity:.4f} | GDR (g-mean): {macro_gdr:.4f}")
print(f"Weighted - Precision: {weighted_prec:.4f} | Recall: {weighted_rec:.4f} | F1: {weighted_f1:.4f} | Specificity: {weighted_specificity:.4f} | GDR (g-mean): {weighted_gdr:.4f}")
print(f"MCC: {mcc:.4f}")
print("Per-class metrics saved to CSV and printed above.")
```

```python
# Single notebook cell: Generate Grad-CAM images (5 per class) and display in grid (5 per row)
# -----------------------------------------------------------------------------
import tensorflow as tf
import numpy as np
import cv2
import os
import random
from pathlib import Path
from tqdm import tqdm
import matplotlib.pyplot as plt
import math

# ---------------------- USER CONFIG ----------------------
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
preprocess_fn = tf.keras.applications.densenet.preprocess_input

CROP_NAME = "Grape"   # e.g. "Apple", "Tomato"
PRE_TRAINED_MODEL = "DenseNet121_TrainTestVal"
OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")   # where tflite + weights .npy live and where outputs will be written
# Local
# DATASET_ROOT = Path("../Datasets/plant-village-dataset-updated-aggregated")
# Colab
DATASET_ROOT = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated")
IMG_SIZE = (224, 224)   # model input size
IMAGES_PER_CLASS = 5    # how many random images per class to create heatmaps for
SCALE_0_1 = False       # set True if model expects inputs scaled to [0,1]; False to use 0..255 floats
VERBOSE = True
# ---------------------------------------------------------

# Derived paths
DATASET_PATH = DATASET_ROOT / CROP_NAME
TFLITE_PATH = OUT_DIR / f"{CROP_NAME.lower()}_two_output_{PRE_TRAINED_MODEL}.tflite"
WEIGHTS_PATH = OUT_DIR / f"{CROP_NAME.lower()}_classifier_weights_combined_{PRE_TRAINED_MODEL}.npy"
BIAS_PATH = OUT_DIR / f"{CROP_NAME.lower()}_classifier_bias_combined_{PRE_TRAINED_MODEL}.npy"
LABELS_PATH = Path(f"{CROP_NAME.lower()}_disease_labels.txt")
SAVE_DIR = OUT_DIR / "gradcam" / CROP_NAME
SAVE_DIR.mkdir(parents=True, exist_ok=True)

# Sanity checks
assert DATASET_PATH.exists(), f"Dataset folder not found: {DATASET_PATH}"
assert TFLITE_PATH.exists(), f"TFLite file not found: {TFLITE_PATH}"
assert WEIGHTS_PATH.exists(), f"Weights file not found: {WEIGHTS_PATH}"
assert LABELS_PATH.exists(), f"Labels file not found: {LABELS_PATH}"

# Load labels
with open(LABELS_PATH, "r") as f:
    class_names = [ln.strip() for ln in f if ln.strip()]
num_classes = len(class_names)
if VERBOSE:
    print(f"Loaded {num_classes} labels: {class_names}")

# Load classifier weights & bias
W = np.load(WEIGHTS_PATH)   # expected shape (channels, num_classes) if GAP -> Dense
b = None
if BIAS_PATH.exists():
    b = np.load(BIAS_PATH)

if VERBOSE:
    print("Classifier weight shape:", W.shape, " bias shape:", None if b is None else b.shape)

# Load TFLite interpreter
interpreter = tf.lite.Interpreter(model_path=str(TFLITE_PATH))
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
if VERBOSE:
    print("TFLite Input details:", input_details)
    print("TFLite Output details:", output_details)

# Identify which output index is feature maps and which is probabilities:
feat_idx = None
prob_idx = None
for od in output_details:
    shape = od["shape"]
    if len(shape) == 4:
        feat_idx = od["index"]
    elif len(shape) == 2:
        prob_idx = od["index"]

if feat_idx is None or prob_idx is None:
    raise RuntimeError("Couldn't identify feature-map and prob outputs from TFLite. Check model outputs.")

if VERBOSE:
    print("Using feature-output index:", feat_idx, "prob-output index:", prob_idx)

# Helper: image preproc (assumes model contains preprocess step, so pass raw 0..255 floats)
def preprocess_image(path: Path):
    img_bgr = cv2.imread(str(path))
    if img_bgr is None:
        raise ValueError(f"Failed to load image {path}")
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    resized = cv2.resize(img_rgb, IMG_SIZE)
    arr = resized.astype(np.float32)
    if SCALE_0_1:
        arr = arr / 255.0
    else:
        arr = preprocess_fn(arr)
    return arr, resized  # arr used as model input, resized is used for overlay visualization (RGB uint8)

# Helper: compute grad-cam heatmap from feature maps and classifier weights
def compute_gradcam_from_featuremaps(feature_maps: np.ndarray, class_id: int, weights: np.ndarray):
    """
    feature_maps: (1, H, W, C) float32
    weights: (C, num_classes) or (num_classes, C)
    returns heatmap HxW normalized [0,1]
    """
    fm = feature_maps[0]   # H x W x C
    # find correct weight vector for class
    if weights.shape[0] == fm.shape[2] and weights.shape[1] == num_classes:
        w_k = weights[:, class_id]    # (C,)
    elif weights.shape[1] == fm.shape[2] and weights.shape[0] == num_classes:
        w_k = weights[class_id, :]
    else:
        w_try = weights.T
        if w_try.shape[0] == fm.shape[2]:
            w_k = w_try[:, class_id]
        else:
            raise ValueError(f"Classifier weights shape {weights.shape} incompatible with feature channels {fm.shape[2]}")

    heatmap = np.tensordot(fm, w_k, axes=([2], [0]))   # shape (H, W)
    heatmap = np.maximum(heatmap, 0)
    if np.max(heatmap) > 0:
        heatmap = heatmap / np.max(heatmap)
    else:
        heatmap = np.zeros_like(heatmap)
    return heatmap.astype(np.float32)

# Helper: overlay heatmap on image
def overlay_heatmap_on_image(orig_rgb, heatmap, alpha=0.45, colormap=cv2.COLORMAP_JET):
    hm_u8 = np.uint8(255 * heatmap)
    hm_color = cv2.applyColorMap(hm_u8, colormap)         # BGR
    hm_color = cv2.cvtColor(hm_color, cv2.COLOR_BGR2RGB)  # to RGB
    hm_resized = cv2.resize(hm_color, (orig_rgb.shape[1], orig_rgb.shape[0]))
    overlay = (0.5 * orig_rgb + alpha * hm_resized).astype(np.uint8)
    return overlay

# Collect up to IMAGES_PER_CLASS random images per class
samples = []
for cls in class_names:
    folder = DATASET_PATH / cls
    if not folder.exists():
        print(f"⚠️ Missing folder {folder}, skipping")
        continue
    files = [Path(folder) / f for f in os.listdir(folder) if f.lower().endswith(('.jpg','.jpeg','.png','.bmp','.tif','.tiff'))]
    if len(files) == 0:
        continue
    chosen = random.sample(files, k=min(IMAGES_PER_CLASS, len(files)))
    for p in chosen:
        samples.append((cls, p))

if VERBOSE:
    print(f"Selected {len(samples)} total images ({IMAGES_PER_CLASS} per class where available).")

# Run tflite inference, compute heatmaps, save overlays
records = []
for cls, img_path in tqdm(samples, desc="Generating Grad-CAM", ncols=100):
    img_arr, resized_rgb = preprocess_image(img_path)
    input_tensor = np.expand_dims(img_arr, axis=0).astype(np.float32)

    interpreter.set_tensor(input_details[0]['index'], input_tensor)
    interpreter.invoke()
    feat = interpreter.get_tensor(feat_idx)   # (1,Hf,Wf,C)
    probs = interpreter.get_tensor(prob_idx)  # (1,num_classes)
    probs = np.asarray(probs).astype(np.float32)[0]
    pred_idx = int(np.argmax(probs))
    pred_label = class_names[pred_idx] if pred_idx < len(class_names) else str(pred_idx)
    confidence = float(probs[pred_idx])

    heatmap = compute_gradcam_from_featuremaps(feat, pred_idx, W)
    overlay = overlay_heatmap_on_image(resized_rgb, heatmap, alpha=0.5)

    class_folder = SAVE_DIR / cls
    class_folder.mkdir(parents=True, exist_ok=True)

    base_name = img_path.stem
    out_heatmap_path = class_folder / f"{base_name}_heatmap.png"
    out_overlay_path = class_folder / f"{base_name}_overlay.png"
    out_orig_path = class_folder / f"{base_name}_orig.png"

    cv2.imwrite(str(out_heatmap_path), (255 * cv2.resize(heatmap, (IMG_SIZE[0], IMG_SIZE[1]))).astype(np.uint8))
    cv2.imwrite(str(out_overlay_path), cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))
    cv2.imwrite(str(out_orig_path), cv2.cvtColor(resized_rgb, cv2.COLOR_RGB2BGR))

    records.append({
        "true_class": cls,
        "image_path": str(img_path),
        "predicted_label": pred_label,
        "confidence": confidence,
        "overlay": str(out_overlay_path),
        "orig": str(out_orig_path),
        "heatmap": str(out_heatmap_path)
    })

print(f"\nSaved {len(records)} Grad-CAM images to: {SAVE_DIR}")

# ------------------ DISPLAY: grid with 5 images per row ------------------
# ------------------ DISPLAY: grid with 5 images per row, no horizontal scroll, dark bold titles ------------------
if len(records) == 0:
    print("No records to display.")
else:
    per_row = 5                       # show exactly 5 images per row
    n = len(records)
    rows = math.ceil(n / per_row)

    # Prevent horizontal scrolling by capping figure width (in inches).
    # Increase/decrease max_width_inch if you know your display width.
    max_width_inch = 15
    cell_size_inch = 2.8              # size per image cell (adjust if you want larger/smaller)
    fig_w = min(per_row * cell_size_inch, max_width_inch)
    fig_h = rows * cell_size_inch

    fig, axes = plt.subplots(rows, per_row, figsize=(fig_w, fig_h))
    # ensure axes is 2D array even if rows==1
    if rows == 1:
        axes = np.expand_dims(axes, axis=0)
    axes = np.array(axes).reshape(rows, per_row)

    for idx, rec in enumerate(records):
        r = idx // per_row
        c = idx % per_row
        ax = axes[r, c]
        img_bgr = cv2.imread(rec["overlay"])
        if img_bgr is None:
            img_bgr = cv2.imread(rec["orig"])  # fallback
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
        ax.imshow(img_rgb)
        title = f"T:{rec['true_class']}  P:{rec['predicted_label']} ({rec['confidence']*100:.1f}%)"
        ax.set_title(title, fontsize=9, color='black', fontweight='bold')   # dark bold font
        ax.axis("off")

    # hide unused subplots
    for idx in range(n, rows * per_row):
        r = idx // per_row
        c = idx % per_row
        axes[r, c].axis("off")

    plt.tight_layout()
    plt.show()

```
