#### Leaf CLass Identifier - Test and Val

```python
import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.metrics import classification_report, confusion_matrix

# ==============================
# CONFIG
# ==============================
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

# Point to dataset where each top-level folder is a crop (Apple, Corn, Grape, ...)
#DATASET_PATH = Path("../Datasets/plant-village-dataset-updated-aggregated")
DATASET_PATH = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated")

IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 15
PRE_TRAINED_MODEL = "MobileNetV3Large"

# ==============================
# DATA LOADER (each folder under DATASET_PATH is a class: Apple, Corn, ...)
# ==============================
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    DATASET_PATH,
    validation_split=0.2,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    DATASET_PATH,
    validation_split=0.2,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

class_names = train_ds.class_names
num_classes = len(class_names)

print(f"✅ Training leaf-TYPE classifier with {num_classes} classes")
print("Classes (leaf types):", class_names)

# Save class labels for React Native app
with open("leaf_type_labels.txt", "w") as f:
    for name in class_names:
        f.write(name + "\n")

# ==============================
# DATA AUGMENTATION + PIPELINE
# ==============================
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
    tf.keras.layers.RandomContrast(0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
    tf.keras.layers.GaussianNoise(0.05),
    tf.keras.layers.RandomBrightness(factor=0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
])

AUTOTUNE = tf.data.AUTOTUNE

train_ds = (
    train_ds
    .cache()
    .shuffle(1000)
    .map(lambda x, y: (data_augmentation(x, training=True), y),
         num_parallel_calls=AUTOTUNE)
    .prefetch(buffer_size=AUTOTUNE)
)

val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)

# ==============================
# MODEL BUILDING - MobileNetV3 Large
# ==============================
preprocess = tf.keras.applications.mobilenet_v3.preprocess_input

base_model = tf.keras.applications.MobileNetV3Large(
    input_shape=IMG_SIZE + (3,),
    include_top=False,
    weights="imagenet"
)
base_model.trainable = False  # freeze backbone initially

inputs = tf.keras.Input(shape=IMG_SIZE + (3,), name="input_image")
# keep preprocess inside model so TFLite can accept raw RGB [0..255] input
x = tf.keras.layers.Lambda(lambda img: preprocess(img), name="preprocess")(inputs)
x = base_model(x, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dropout(0.3)(x)
outputs = tf.keras.layers.Dense(num_classes, activation="softmax", name="predictions")(x)

model = tf.keras.Model(inputs, outputs, name=f"{PRE_TRAINED_MODEL}_leaf_type_classifier")

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# ==============================
# TRAINING
# ==============================
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss", patience=5, restore_best_weights=True
)

checkpoint = tf.keras.callbacks.ModelCheckpoint(
    f"best_leaf_type_classifier_{PRE_TRAINED_MODEL}.keras",
    save_best_only=True,
    monitor="val_accuracy",
    mode="max"
)

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# Fine-tuning: unfreeze some top layers of base_model
base_model.trainable = True
fine_tune_at = int(len(base_model.layers) * 0.75)
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_fine = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# ==============================
# PLOTS: Accuracy & Loss
# ==============================
plt.figure(figsize=(10, 6))
plt.plot(history.history.get('accuracy', []) + history_fine.history.get('accuracy', []), label="Training Accuracy")
plt.plot(history.history.get('val_accuracy', []) + history_fine.history.get('val_accuracy', []), label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Leaf TYPE Classifier - Training & Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(history.history.get('loss', []) + history_fine.history.get('loss', []), label="Training Loss")
plt.plot(history.history.get('val_loss', []) + history_fine.history.get('val_loss', []), label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Leaf TYPE Classifier - Training & Validation Loss")
plt.legend()
plt.grid(True)
plt.show()

# ==============================
# EVALUATION: Classification Report & Confusion Matrix
# ==============================
y_true = []
y_pred = []

for batch_images, batch_labels in val_ds:
    preds = model.predict(batch_images, verbose=0)
    y_true.extend(batch_labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

print("📊 Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred, labels=range(num_classes))
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Leaf TYPE Classifier - Confusion Matrix")
plt.show()

# ==============================
# EXPORT TFLITE (float32 model expecting raw RGB [0..255] inputs)
# ==============================
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

tflite_file = f"leaf_type_classifier_{PRE_TRAINED_MODEL}.tflite"
with open(tflite_file, "wb") as f:
    f.write(tflite_model)

print(f"✅ TFlite model saved: {tflite_file}")
```

#### Leaf Disease Detection MobileNetV3Large - Train, Test, and Val

```python
import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.metrics import classification_report, confusion_matrix
import os
from collections import Counter
import pandas as pd

# ==============================
# CONFIG
# ==============================
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

DATASET_PATH = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/plant-village-dataset-updated-aggregated")
OUT_DIR = Path("/content/drive/MyDrive/Masters_LJMU/Datasets/exported_models")
OUT_DIR.mkdir(parents=True, exist_ok=True)

IMG_SIZE = (224, 224)
BATCH_SIZE = 64
EPOCHS = 5
PRE_TRAINED_MODEL = "MobileNetV3Large_TrainTestVal"
DROPOUT = 0.3
LR_RATE = 1e-4

# Data Loader helpers
def count_per_class(file_paths, class_names):
    labels = [Path(f).parent.name for f in file_paths]
    counts = Counter(labels)
    return {cls: counts.get(cls, 0) for cls in class_names}

# collect all image files (for counting checks)
all_files = []
for root, _, files in os.walk(DATASET_PATH):
    for f in files:
        if f.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp")):
            all_files.append(Path(root) / f)
N = len(all_files)
if N == 0:
    raise RuntimeError(f"No images found under {DATASET_PATH}")

train_count = int(N * 0.70)
val_count = int(N * 0.15)
test_count = N - train_count - val_count

print(f"Total images: {N} -> train: {train_count}, val: {val_count}, test: {test_count}")

# use validation_split to get training + rest (30%)
rest_split = 0.3

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    DATASET_PATH,
    validation_split=rest_split,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True
)

rest_ds = tf.keras.preprocessing.image_dataset_from_directory(
    DATASET_PATH,
    validation_split=rest_split,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True
)

class_names = train_ds.class_names
num_classes = len(class_names)

# Save class labels for React Native app (save inside OUT_DIR)
labels_path = OUT_DIR / "leaf_type_labels.txt"
with open(labels_path, "w") as f:
    for name in class_names:
        f.write(name + "\n")
print(f"Saved class labels to: {labels_path}")

# ---------------------------
# NEW: Deterministic per-class Train/Val/Test counts (minimal change)
# ---------------------------
IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp"}

rows = []
rng = random.Random(SEED)  # deterministic shuffle per class

for cls_dir in sorted([p for p in DATASET_PATH.iterdir() if p.is_dir()]):
    files = [p for p in cls_dir.rglob("*") if p.suffix.lower() in IMG_EXTS]
    files = [str(p) for p in files]  # convert to strings for shuffling
    rng.shuffle(files)
    n = len(files)
    tr = int(n * 0.70)
    va = int(n * 0.15)
    te = n - tr - va
    rows.append({"crop": cls_dir.name, "Train": tr, "Val": va, "Test": te, "Total": n})

df_counts = pd.DataFrame(rows).set_index("crop").sort_index()
print("\nPer-class Train/Val/Test counts:\n")
print(df_counts)
csv_out = OUT_DIR / "per_class_train_val_test_counts.csv"
df_counts.to_csv(csv_out)
print(f"\nSaved per-class counts to: {csv_out}")

# ---------------------------
# Continue with your pipeline (unchanged)
# ---------------------------

# Print the count of images per class for train/rest splits (approx)
train_files = train_ds.file_paths
rest_files = rest_ds.file_paths
val_files = rest_files[:val_count]
test_files = rest_files[val_count:val_count + test_count]

# (The rest of your original script continues from here — unchanged)
# Flatten rest_ds and split deterministically by counts (rest_ds contains 30% of images)
rest_ds = rest_ds.unbatch()
val_ds = rest_ds.take(val_count)
test_ds = rest_ds.skip(val_count).take(test_count)

# Rebatch val & test
val_ds = val_ds.batch(BATCH_SIZE)
test_ds = test_ds.batch(BATCH_SIZE)

print("\n--------------------")
print(f"✅ Training {DATASET_PATH} model with {num_classes} classes")
print("Classes:", class_names)
train_batches = tf.data.experimental.cardinality(train_ds).numpy()
print(f"Train dataset: {train_batches} batches (batch_size={BATCH_SIZE})")

# ==============================
# DATA AUGMENTATION (as you requested)
# ==============================
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
    tf.keras.layers.RandomContrast(0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
    tf.keras.layers.GaussianNoise(0.05),
    tf.keras.layers.RandomBrightness(factor=0.2),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
])

AUTOTUNE = tf.data.AUTOTUNE

# Pipelines

# TRAINING pipeline: cache -> shuffle -> augment -> repeat -> prefetch
train_ds = (
    train_ds
    .cache()
    .shuffle(1000)
    .map(lambda x, y: (data_augmentation(x, training=True), y),
         num_parallel_calls=AUTOTUNE)
    .repeat()   # <-- ADD repeat here so fit() can run many epochs
    .prefetch(buffer_size=AUTOTUNE)
)

# VALIDATION for training (use a repeated version for validation during fit)
val_ds_for_fit = (
    val_ds
    .cache()
    .repeat()
    .prefetch(buffer_size=AUTOTUNE)
)

# VAL pipeline used for final evaluation remains finite (val_ds)
# TEST pipeline remains finite
test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE)

# compute steps for training loops
steps_per_epoch = max(1, train_count // BATCH_SIZE)
validation_steps = max(1, val_count // BATCH_SIZE)
print(f"steps_per_epoch={steps_per_epoch}, validation_steps={validation_steps}")

# ==============================
# MODEL BUILDING - MobileNetV3 Large with preprocessing included
# ==============================
inputs = tf.keras.layers.Input(shape=IMG_SIZE + (3,), name="input_image")

base_model = tf.keras.applications.MobileNetV3Large(
    input_shape=IMG_SIZE + (3,),
    include_top=False,
    weights="imagenet",
    include_preprocessing=True
)
base_model.trainable = False  # freeze backbone initially

x = base_model(inputs, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(256, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(LR_RATE))(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dropout(DROPOUT)(x)
outputs = tf.keras.layers.Dense(num_classes, activation="softmax", name="predictions")(x)

model = tf.keras.Model(inputs, outputs, name=f"{PRE_TRAINED_MODEL}_leaf_type_classifier")

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LR_RATE),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# ==============================
# TRAINING
# ==============================
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss", patience=10, restore_best_weights=True
)

checkpoint_path = OUT_DIR / f"best_leaf_type_classifier_{PRE_TRAINED_MODEL}.keras"
checkpoint = tf.keras.callbacks.ModelCheckpoint(
    str(checkpoint_path),
    save_best_only=True,
    monitor="val_accuracy",
    mode="max"
)

history = model.fit(
    train_ds,
    validation_data=val_ds_for_fit,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# Fine-tuning: unfreeze top layers
base_model.trainable = True
fine_tune_at = int(len(base_model.layers) * 0.75)
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_fine = model.fit(
    train_ds,
    validation_data=val_ds_for_fit,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCHS,
    callbacks=[early_stop, checkpoint]
)

# (rest of your evaluation / tflite export unchanged)
# Evaluate on validation
results = model.evaluate(val_ds)
print("Metrics - ")
for name, value in zip(model.metrics_names, results):
    print(f"{name}: {value:.4f}")

# Plots: Accuracy & Loss
plt.figure(figsize=(10, 6))
plt.plot(history.history.get('accuracy', []) + history_fine.history.get('accuracy', []), label="Training Accuracy")
plt.plot(history.history.get('val_accuracy', []) + history_fine.history.get('val_accuracy', []), label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title(f"Training & Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(history.history.get('loss', []) + history_fine.history.get('loss', []), label="Training Loss")
plt.plot(history.history.get('val_loss', []) + history_fine.history.get('val_loss', []), label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title(f"Training & Validation Loss")
plt.legend()
plt.grid(True)
plt.show()

# Validation Evaluation: Classification Report & Confusion Matrix on VAL set
# ==============================
y_true = []
y_pred = []

for batch_images, batch_labels in val_ds:
    preds = model.predict(batch_images, verbose=0)
    y_true.extend(batch_labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

# Safety: ensure lengths align
if len(class_names) != (int(max(max(y_true, default=0), max(y_pred, default=0))) + 1):
    print("Warning: class_names length does not match labels. Recomputing class_names from folders.")
    class_names = sorted([d.name for d in DATASET_PATH.iterdir() if d.is_dir()])
    num_classes = len(class_names)
    print("New class_names:", class_names)

y_true = np.array(y_true, dtype=np.int64)
y_pred = np.array(y_pred, dtype=np.int64)

acc = float(np.mean(y_true == y_pred))
print(f"\n Overall accuracy: {acc*100:.2f}%")

print("Validation Classification Report:")
print(classification_report(y_true, y_pred,labels=list(range(num_classes)), target_names=class_names, zero_division=0))

cm = confusion_matrix(y_true, y_pred, labels=range(len(class_names)))
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"Validation Confusion Matrix")
plt.show()

# Testing Evaluation: classification report & confusion matrix on TEST set
y_true_test = []
y_pred_test = []

for batch_images, batch_labels in test_ds:
    preds = model.predict(batch_images, verbose=0)
    y_true_test.extend(batch_labels.numpy())
    y_pred_test.extend(np.argmax(preds, axis=1))

y_true_test = np.array(y_true_test, dtype=np.int64)
y_pred_test = np.array(y_pred_test, dtype=np.int64)

acc = float(np.mean(y_true_test == y_pred_test))
print(f"\nOverall accuracy: {acc*100:.2f}%")

print("Test Classification Report:")
print(classification_report(y_true_test, y_pred_test, target_names=class_names, zero_division=0))

cm_test = confusion_matrix(y_true_test, y_pred_test, labels=range(len(class_names)))
plt.figure(figsize=(8, 6))
sns.heatmap(cm_test, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"Test Confusion Matrix")
plt.show()

# Convert wrapper to TFLite; try normal, fallback to SELECT_TF_OPS if needed
TFLITE_PATH = OUT_DIR / f"leaf_classifier_output_{PRE_TRAINED_MODEL}.tflite"
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# try conversion; if converter needs TF ops, convert with SELECT_TF_OPS
try:
    tflite_model = converter.convert()
    TFLITE_PATH.write_bytes(tflite_model)
    print(f"✅ TFLite (two-output) saved to: {TFLITE_PATH}")
except Exception as e:
    print("⚠️ First conversion failed — attempting with SELECT_TF_OPS. Error:", e)
    converter.target_spec.supported_ops = [
        tf.lite.OpsSet.TFLITE_BUILTINS,
        tf.lite.OpsSet.SELECT_TF_OPS
    ]
    tflite_model = converter.convert()
    TFLITE_PATH.write_bytes(tflite_model)
    print(f"✅ TFLite (output, SELECT_TF_OPS) saved to: {TFLITE_PATH}")

print("Done. Files in:", OUT_DIR.resolve())

```
